```python
from collections import defaultdict

# 입력 받기
n = int(input())

# (요일, 시간) -> 참석 가능한 학생 집합
availability = defaultdict(set)

for _ in range(n):
    parts = input().split()
    name = parts[0]
    day = parts[1]
    x = int(parts[2])
    hours = [int(parts[3 + i]) for i in range(x)]

    # 각 시간에 대해 학생 추가
    for hour in hours:
        availability[(day, hour)].add(name)

# 가장 많은 학생이 참석 가능한 시간 찾기
max_students = 0
best_day = ""
best_hour = 0

for (day, hour), students in availability.items():
    if len(students) > max_students:
        max_students = len(students)
        best_day = day
        best_hour = hour

# 시간 포맷팅 (zero-padded)
time_str = f"{best_hour:02d}:00"

print(f"Your professor should host office hours {best_day} @ {time_str}")
```
