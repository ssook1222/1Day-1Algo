```python
# 입력 받기
a = int(input())  # X사의 1리터당 요금
b = int(input())  # Y사의 기본요금
c = int(input())  # Y사의 기본요금 적용 상한
d = int(input())  # Y사의 1리터당 추가요금
p = int(input())  # 한 달간 사용량

# X사 요금 계산
x_cost = a * p

# Y사 요금 계산
if p <= c:
    y_cost = b
else:
    y_cost = b + (p - c) * d

# 최소 요금 출력
min_cost = min(x_cost, y_cost)
print(min_cost)
```
