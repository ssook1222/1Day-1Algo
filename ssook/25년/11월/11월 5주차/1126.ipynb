## BoJ

### 13063

```python
while True:
    n, m, k = map(int, input().split())

    # 종료 조건
    if n == 0 and m == 0 and k == 0:
        break

    # 과반수를 위해 필요한 최소 표수
    votes_needed = (n // 2) + 1

    # 이미 확보한 표수 (보수당)
    votes_secured = m

    # 추가로 필요한 표수
    additional_votes = votes_needed - votes_secured

    # 무소속 의원 수
    independents = n - m - k

    # 추가 표가 필요 없으면 0
    if additional_votes <= 0:
        print(0)
    # 무소속 의원이 충분하면 필요한 수 출력
    elif additional_votes <= independents:
        print(additional_votes)
    # 무소속 의원이 부족하면 -1
    else:
        print(-1)
```
