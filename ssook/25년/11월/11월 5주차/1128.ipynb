```python
def calculate_games(n, m, k, b):
    """주어진 B 값에 대해 전체 경기 수를 계산"""
    a = k * b

    same_region_games = n * m * (m - 1) // 2 * a
    diff_region_games = n * (n - 1) // 2 * m * m * b

    return same_region_games + diff_region_games

def solve(n, m, k, d):
    max_b = d // (n * m * (m - 1) // 2 * k + n * (n - 1) // 2 * m * m) + 1
    max_b = max(max_b, 1000)  # 최소한 충분히 큰 값으로

    best_games = -1

    b = 1
    while True:
        games = calculate_games(n, m, k, b)

        if games <= d:
            best_games = games
            b += 1
        else:
            break

        if b > max_b:
            break

    return best_games

# 입력 처리
t = int(input())
for _ in range(t):
    n, m, k, d = map(int, input().split())
    result = solve(n, m, k, d)
    print(result)
```
