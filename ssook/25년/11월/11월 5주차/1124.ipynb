## BoJ

### 30045

```python
n = int(input())
count = 0

for _ in range(n):
    s = input()
    # 문장에 '01' 또는 'OI'가 포함되어 있는지 확인
    if '01' in s or 'OI' in s:
        count += 1

print(count)
```
