```python
def check_file_transfer():
    n = int(input())
    valid_count = 0

    for _ in range(n):
        filename = input().strip()

        # 점(.)이 정확히 하나 있는지 확인
        dot_count = filename.count('.')
        if dot_count != 1:
            continue

        # 점을 기준으로 분리
        parts = filename.split('.')
        name_part = parts[0]  # filename 부분
        ext_part = parts[1]   # extension 부분

        # filename 부분 검증
        # 1. 비어있지 않아야 함
        # 2. 최대 8자
        # 3. 라틴 소문자만 포함
        if len(name_part) == 0 or len(name_part) > 8:
            continue
        if not name_part.isalpha():
            continue

        # extension 부분 검증
        # 1. 비어있지 않아야 함
        # 2. 최대 3자
        # 3. 라틴 소문자만 포함
        if len(ext_part) == 0 or len(ext_part) > 3:
            continue
        if not ext_part.isalpha():
            continue

        # 모든 조건을 만족하면 유효한 파일
        valid_count += 1

    print(valid_count)

# 실행
check_file_transfer()
```
