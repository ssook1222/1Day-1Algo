```python
def solve_isbn():
    isbn = input().strip()

    def calculate_isbn_sum(isbn_str):
        """ISBN 문자열의 가중합을 계산"""
        total = 0
        for i, char in enumerate(isbn_str):
            weight = 10 - i
            if char == 'X':
                digit_value = 10
            else:
                digit_value = int(char)
            total += weight * digit_value
        return total

    def is_valid_isbn(isbn_str):
        """ISBN이 유효한지 확인 (11로 나누어떨어지는지)"""
        return calculate_isbn_sum(isbn_str) % 11 == 0

    # '?' 위치 찾기
    question_pos = isbn.index('?')

    # 0~9를 시도
    for digit in range(10):
        test_isbn = isbn[:question_pos] + str(digit) + isbn[question_pos + 1:]
        if is_valid_isbn(test_isbn):
            print(digit)
            return

    # X를 시도 (마지막 자리인 경우만)
    if question_pos == 9:  # 마지막 자리
        test_isbn = isbn[:question_pos] + 'X' + isbn[question_pos + 1:]
        if is_valid_isbn(test_isbn):
            print('X')
            return

    # 유효한 숫자가 없는 경우
    print(-1)

# 실행
solve_isbn()
```
