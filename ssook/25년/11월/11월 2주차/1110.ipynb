```python
def solve_sushi():
    n = int(input())
    customers = []
    for i in range(n):
        menu = input().strip()
        customers.append(menu)

    # 각 종류의 스시가 총 몇 개 필요한지 계산
    sushi_count = {}

    for customer_menu in customers:
        for sushi in customer_menu:
            sushi_count[sushi] = sushi_count.get(sushi, 0) + 1

    # 필요한 스시의 총 개수가 답
    total_sushi_needed = sum(sushi_count.values())

    return total_sushi_needed

# 실행
result = solve_sushi()
print(result)
```
