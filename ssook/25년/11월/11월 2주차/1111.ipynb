```python
def solve_dickensian():
    word = input().strip()

    # 왼손과 오른손 글자 정의
    left_hand = set("qwertasdfgzxcvb")
    right_hand = set("yuiophjklnm")

    # 첫 번째 글자가 어느 손인지 확인
    if word[0] in left_hand:
        current_hand = "left"
    elif word[0] in right_hand:
        current_hand = "right"
    else:
        # 예외 상황 (문제에서는 a-z만 주어진다고 했으므로 발생하지 않음)
        print("no")
        return

    # 두 번째 글자부터 확인
    for i in range(1, len(word)):
        char = word[i]

        if char in left_hand:
            next_hand = "left"
        elif char in right_hand:
            next_hand = "right"
        else:
            # 예외 상황
            print("no")
            return

        # 연속된 두 글자가 같은 손이면 Dickensian이 아님
        if current_hand == next_hand:
            print("no")
            return

        # 다음 글자를 위해 현재 손 업데이트
        current_hand = next_hand

    # 모든 글자가 번갈아가며 타이핑되면 Dickensian
    print("yes")

# 실행
solve_dickensian()
```
