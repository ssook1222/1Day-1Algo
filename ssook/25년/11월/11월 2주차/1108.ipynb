```python
def find_missing_obstacles():
    # 첫 번째 줄: N(총 장애물 수), Y(마리오가 발견했다고 한 장애물 수)
    N, Y = map(int, input().split())

    # 마리오가 찾은 장애물들을 저장할 set (중복 제거를 위해)
    found_obstacles = set()

    # Y개의 장애물 번호 입력받기
    for _ in range(Y):
        obstacle = int(input())
        found_obstacles.add(obstacle)

    # 전체 장애물 set 생성 (0부터 N-1까지)
    all_obstacles = set(range(N))

    # 놓친 장애물들 = 전체 장애물 - 찾은 장애물들
    missed_obstacles = all_obstacles - found_obstacles

    # 놓친 장애물들을 오름차순으로 정렬하여 출력
    for obstacle in sorted(missed_obstacles):
        print(obstacle)

    # 마리오가 실제로 찾은 서로 다른 장애물의 개수 출력
    print(f"Mario got {len(found_obstacles)} of the dangerous obstacles.")

# 실행
find_missing_obstacles()
```
