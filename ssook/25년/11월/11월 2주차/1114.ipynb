```python
def solve_real_estate():
    # 입력 읽기
    n, c = map(int, input().split())

    total_area = 0
    bedroom_area = 0
    balcony_area = 0
    other_rooms_area = 0

    # 각 방의 정보 처리
    for _ in range(n):
        line = input().split()
        area = int(line[0])
        room_type = line[1]

        # 총 면적에 추가
        total_area += area

        # 방 타입별 처리
        if room_type == "bedroom":
            bedroom_area += area
            other_rooms_area += area
        elif room_type == "balcony":
            balcony_area += area
        else:  # bathroom, kitchen, other
            other_rooms_area += area

    # 감소된 총 면적 계산
    # 발코니를 제외한 모든 방의 면적 + 발코니 면적의 절반
    reduced_total_area = other_rooms_area + balcony_area * 0.5

    # 아파트 비용 계산
    flat_cost = reduced_total_area * c

    # 출력
    print(total_area)
    print(bedroom_area)
    print(flat_cost)

# 실행
solve_real_estate()
```
