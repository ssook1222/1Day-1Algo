### 7600

```python
while True:
    line = input()

    # '#'이 나오면 종료
    if line == '#':
        break

    # 알파벳만 추출하여 소문자로 변환
    alphabets = set()
    for char in line:
        if char.isalpha():
            alphabets.add(char.lower())

    # 서로 다른 알파벳의 개수 출력
    print(len(alphabets))
```
