```python
def solve_alien_device():
    # 입력 읽기
    n = int(input())
    a = list(map(int, input().split()))
    m = int(input())

    # 각 연산 처리
    for _ in range(m):
        operation = list(map(int, input().split()))
        k = operation[0]  # 연산 종류
        l = operation[1] - 1  # 0-indexed로 변환
        r = operation[2] - 1  # 0-indexed로 변환

        if k == 1:
            # 연산 1: 구간 [l, r]의 모든 원소를 제곱하고 2010으로 나눈 나머지
            for i in range(l, r + 1):
                a[i] = (a[i] * a[i]) % 2010

        elif k == 2:
            # 연산 2: 구간 [l, r]의 합을 출력
            result = sum(a[l:r + 1])
            print(result)

# 실행
solve_alien_device()
```
