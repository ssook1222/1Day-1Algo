```python
n = int(input())
x = int(input())

# n명이 일렬로 서있을 때, 첫 번째와 마지막 사람 사이의 거리
# = (사람 수 - 1) × 사람 사이 거리
max_distance = (n - 1) * x

print(max_distance)
```
