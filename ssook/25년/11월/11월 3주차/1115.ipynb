## BoJ

### 3533

```python
def solve_boolean_function():
    # 입력 읽기
    x = list(map(int, input().split()))

    # x를 0-indexed로 사용하기 위해 (x[0] = x1, x[1] = x2, ...)
    result = 0

    # 모든 쌍 (xi OR xj)을 XOR
    for i in range(10):
        for j in range(i + 1, 10):
            pair_or = x[i] | x[j]  # OR 연산
            result ^= pair_or      # XOR 연산

    # 모든 삼중항 (xi OR xj OR xk)을 XOR
    for i in range(10):
        for j in range(i + 1, 10):
            for k in range(j + 1, 10):
                triplet_or = x[i] | x[j] | x[k]  # OR 연산
                result ^= triplet_or             # XOR 연산

    print(result)

# 실행
solve_boolean_function()
```
