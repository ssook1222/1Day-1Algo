```python
scenario = 1

while True:
    n, d = map(int, input().split())

    if n == 0 and d == 0:
        break

    # 제거된 옷 정보
    becs_removed = int(input())
    cas_removed = int(input())

    # 초기 옷 리스트 (1부터 n까지)
    becs_outfits = list(range(1, n + 1))
    cas_outfits = list(range(1, n + 1))

    # Becs가 옷을 버린 경우 (왼쪽에서 세는 번호)
    if becs_removed != 0:
        becs_outfits.pop(becs_removed - 1)

    # Cas가 옷을 버린 경우 (오른쪽에서 세는 번호)
    if cas_removed != 0:
        # 오른쪽에서 cas_removed번째 = 왼쪽에서 (n - cas_removed + 1)번째
        cas_outfits.pop(n - cas_removed)

    print(f"Scenario {scenario}")

    for day in range(1, d + 1):
        becs_choice, cas_choice = map(int, input().split())

        # Becs의 선택 (왼쪽에서 세기)
        becs_outfit = becs_outfits[becs_choice - 1]

        # Cas의 선택 (오른쪽에서 세기)
        cas_outfit = cas_outfits[len(cas_outfits) - cas_choice]

        if becs_outfit == cas_outfit:
            print(f"Day {day} ALERT")
        else:
            print(f"Day {day} OK")

    scenario += 1
```
