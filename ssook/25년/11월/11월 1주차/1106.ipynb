```python
def calculate_wallet_value():
    # 대통령 성과 지폐 가치 매핑
    president_values = {
        'Franklin': 100,
        'Grant': 50,
        'Jackson': 20,
        'Hamilton': 10,
        'Lincoln': 5,
        'Washington': 1
    }

    n = int(input())

    for _ in range(n):
        line = input().strip()
        presidents = line.split()

        total_value = 0
        for president in presidents:
            if president in president_values:
                total_value += president_values[president]

        print(f"${total_value}")

# 실행
calculate_wallet_value()
```
