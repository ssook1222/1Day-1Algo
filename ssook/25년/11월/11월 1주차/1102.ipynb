```python
def solve_hiarc_emoji():
    """
    문자열 S에서 HI-ARC 이모지를 최대 몇 개 만들 수 있는지 계산
    각 이모지를 만들려면 H, I, A, R, C가 각각 하나씩 필요
    """

    # 입력 받기
    n = int(input())
    s = input().strip()

    # 필요한 문자들의 개수 세기
    required_chars = ['H', 'I', 'A', 'R', 'C']
    char_count = {}

    # 각 필요한 문자의 개수를 0으로 초기화
    for char in required_chars:
        char_count[char] = 0

    # 문자열에서 각 문자의 개수 세기
    for char in s:
        if char in char_count:
            char_count[char] += 1

    # 만들 수 있는 이모지의 최대 개수는
    # 각 필요한 문자 개수의 최솟값
    max_emojis = min(char_count.values())

    return max_emojis

# 메인 실행
if __name__ == "__main__":
    result = solve_hiarc_emoji()
    print(result)
```
