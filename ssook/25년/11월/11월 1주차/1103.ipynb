```python
def find_southernmost_point():
    n = int(input())
    stations = []

    for _ in range(n):
        x, y = map(int, input().split())
        stations.append((x, y))

    # 가장 남쪽 점을 저장할 변수 (y좌표가 가장 작은 점)
    min_y = float('inf')
    result_x, result_y = 0, 0

    # 각 역에서 가장 남쪽 점 확인
    for x, y in stations:
        if y < min_y:
            min_y = y
            result_x, result_y = x, y

    # 각 선분에서 가장 남쪽 점 확인
    for i in range(n - 1):
        x1, y1 = stations[i]
        x2, y2 = stations[i + 1]

        # 선분의 가장 남쪽 점 찾기
        # 선분이 수직이 아닌 경우, y의 최솟값은 끝점 중 하나
        segment_min_y = min(y1, y2)

        if segment_min_y < min_y:
            min_y = segment_min_y
            # y가 최소인 점의 x좌표 찾기
            if y1 < y2:
                result_x, result_y = x1, y1
            elif y2 < y1:
                result_x, result_y = x2, y2
            else:  # y1 == y2인 경우 (수평선분)
                # 이 경우는 선분 전체가 같은 y값을 가지므로
                # 문제에서 유일한 점이라고 했으므로 고려하지 않아도 됨
                result_x, result_y = x1, y1

    print(result_x, result_y)

# 실행
find_southernmost_point()
```
