## BoJ

### 6553

```python
import sys

def solve():
    try:
        while True:
            line = input().strip()
            if not line:
                break

            coords = list(map(int, line.split()))
            n = coords[0]

            # 나라들의 좌표 추출
            countries = []
            for i in range(1, len(coords), 2):
                x = coords[i]
                y = coords[i + 1]
                countries.append((x, y))

            # 회의 장소들 계산
            meeting_locations = []

            # 인접한 나라들 간의 중점 계산
            for i in range(n):
                # 현재 나라와 다음 나라 (마지막 나라는 첫 번째 나라와 연결)
                current = countries[i]
                next_country = countries[(i + 1) % n]

                # 중점 계산
                mid_x = (current[0] + next_country[0]) / 2.0
                mid_y = (current[1] + next_country[1]) / 2.0

                meeting_locations.append((mid_x, mid_y))

            # 출력
            result = [str(n)]
            for x, y in meeting_locations:
                result.append(f"{x:.6f}")
                result.append(f"{y:.6f}")

            print(" ".join(result))

    except EOFError:
        pass

solve()
```
