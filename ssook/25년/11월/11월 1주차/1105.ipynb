```python
import sys

def count_jokes():
    # 모든 입력을 읽어들임
    text = sys.stdin.read()

    # 전체 텍스트를 줄별로 나누어 처리
    lines = text.split('\n')

    total_count = 0

    # 각 줄에서 "joke" 부분 문자열 개수 세기
    for line in lines:
        # 해당 줄에서 "joke" 개수 세기
        count = line.count("joke")
        total_count += count

    print(total_count)

# 실행
count_jokes()
```
