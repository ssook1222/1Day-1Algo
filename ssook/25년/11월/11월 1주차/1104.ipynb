```python
def solve():
    n = int(input())

    for _ in range(n):
        time_str = input().strip()
        h, m, s = map(int, time_str.split(':'))

        # 각 값을 6비트 이진수로 변환
        h_bin = format(h, '06b')  # 시간을 6비트로
        m_bin = format(m, '06b')  # 분을 6비트로
        s_bin = format(s, '06b')  # 초를 6비트로

        # 3열 방식: 6행 3열
        # 각 열이 시간/분/초를 나타내고, 행우선으로 읽음
        # 1열: h_bin[0], h_bin[1], h_bin[2], h_bin[3], h_bin[4], h_bin[5]
        # 2열: m_bin[0], m_bin[1], m_bin[2], m_bin[3], m_bin[4], m_bin[5]
        # 3열: s_bin[0], s_bin[1], s_bin[2], s_bin[3], s_bin[4], s_bin[5]
        three_col = ""
        for i in range(6):  # 6행
            three_col += h_bin[i] + m_bin[i] + s_bin[i]

        # 3행 방식: 3행 6열
        # 각 행이 시간/분/초를 나타내고, 행우선으로 읽음
        # 1행: h_bin 전체 (6비트)
        # 2행: m_bin 전체 (6비트)
        # 3행: s_bin 전체 (6비트)
        three_row = h_bin + m_bin + s_bin

        print(three_col, three_row)

solve()
```
