## BoJ

### 5032

```python
e, f, c = map(int, input().split())

# 총 빈 병 개수
empty_bottles = e + f

# 마신 탄산 음료 개수
drink_count = 0

# 빈 병이 c개 이상인 동안 반복
while empty_bottles >= c:
    # c개의 빈 병으로 새 병을 몇 개 받을 수 있는지
    new_bottles = empty_bottles // c

    # 새 병을 마심
    drink_count += new_bottles

    # 남은 빈 병 = 교환 후 남은 빈 병 + 마시고 생긴 빈 병
    empty_bottles = empty_bottles % c + new_bottles

print(drink_count)
```
