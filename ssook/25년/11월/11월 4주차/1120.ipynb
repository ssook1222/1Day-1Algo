## BoJ

### 26590

```python
word1, word2 = input().split()

# 더 짧은 단어의 길이 구하기
length = min(len(word1), len(word2))

# 새로운 단어 만들기
mixed_word = ""
for i in range(length):
    if i % 2 == 0:
        # 짝수 인덱스: 첫 번째 단어에서
        mixed_word += word1[i]
    else:
        # 홀수 인덱스: 두 번째 단어에서
        mixed_word += word2[i]

print(mixed_word)
```
