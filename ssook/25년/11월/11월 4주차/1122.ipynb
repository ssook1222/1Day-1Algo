```python
n = int(input())
stores = list(map(int, input().split()))

# 현재 마셔야 할 우유 (0: 딸기, 1: 초코, 2: 바나나)
next_milk = 0
count = 0

# 각 가게를 순서대로 방문
for store in stores:
    # 현재 가게의 우유가 마셔야 할 우유와 같으면
    if store == next_milk:
        count += 1
        # 다음에 마셔야 할 우유로 업데이트
        # 0 -> 1 -> 2 -> 0 -> ... (순환)
        next_milk = (next_milk + 1) % 3

print(count)
```
