### 31995

```python
N = int(input())
M = int(input())

# 대각선으로 이웃하려면 가로와 세로 모두 2칸 이상 필요
if N < 2 or M < 2:
    print(0)
else:
    result = 2 * (N - 1) * (M - 1)
    print(result)
```
