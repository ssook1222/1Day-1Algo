## BoJ

### 11800

```python
T = int(input())

# 주사위 숫자에 대한 별칭
names = {
    1: "Yakk",
    2: "Doh",
    3: "Seh",
    4: "Ghar",
    5: "Bang",
    6: "Sheesh"
}

# 같은 수가 나올 때의 별칭
doubles = {
    1: "Habb Yakk",
    2: "Dobara",
    3: "Dousa",
    4: "Dorgy",
    5: "Dabash",
    6: "Dosh"
}

for case in range(1, T + 1):
    a, b = map(int, input().split())

    # 같은 수가 나온 경우
    if a == b:
        result = doubles[a]
    # 6-5 또는 5-6인 경우
    elif (a == 6 and b == 5) or (a == 5 and b == 6):
        result = "Sheesh Beesh"
    # 일반적인 경우 (큰 수부터)
    else:
        bigger = max(a, b)
        smaller = min(a, b)
        result = f"{names[bigger]} {names[smaller]}"

    print(f"Case {case}: {result}")
```
