## BoJ

### 25053

```python
t = int(input())

for _ in range(t):
    n = int(input())

    # 각 난이도별로 최대 beauty 값을 저장
    max_beauty = {}

    for _ in range(n):
        b, d = map(int, input().split())

        # 해당 난이도에서 가장 아름다운 문제 저장
        if d not in max_beauty:
            max_beauty[d] = b
        else:
            max_beauty[d] = max(max_beauty[d], b)

    # 1부터 10까지 모든 난이도가 있는지 확인
    has_all_difficulties = True
    for difficulty in range(1, 11):
        if difficulty not in max_beauty:
            has_all_difficulties = False
            break

    if not has_all_difficulties:
        print("MOREPROBLEMS")
    else:
        # 각 난이도별 최대 beauty 값을 모두 합산
        total_beauty = sum(max_beauty[d] for d in range(1, 11))
        print(total_beauty)
```
