## BoJ

### 14697

```python
def solve():
    a, b, c, n = map(int, input().split())

    # A인실 x개, B인실 y개, C인실 z개 사용
    # A*x + B*y + C*z = N인 (x, y, z) 존재 여부 확인
    # x, y, z >= 0

    # x의 최대값: N // A
    for x in range(n // a + 1):
        remaining_after_a = n - a * x

        # y의 최대값: remaining_after_a // B
        for y in range(remaining_after_a // b + 1):
            remaining_after_b = remaining_after_a - b * y

            # z를 사용하여 remaining_after_b를 만들 수 있는지 확인
            if remaining_after_b % c == 0:
                # 정확히 나누어떨어지면 가능
                print(1)
                return

    # 불가능
    print(0)

solve()
```
