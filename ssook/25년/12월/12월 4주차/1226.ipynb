```python
H, x = map(int, input().split())

MOD = 10**9 + 7
total = 0

# i+1번째 줄이 높이 i cm에서 만들어진 스노우볼 개수
for height in range(1, H + 1):
    c = int(input())

    # height에서 만들어진 스노우볼이 내려올 거리
    # 높이 height에서 높이 0까지 내려오므로 distance = height
    distance = height

    # 최종 크기 = x^distance
    if c > 0:
        final_size = pow(x, distance, MOD)
        total = (total + c * final_size) % MOD

print(total)
```
