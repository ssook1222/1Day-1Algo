```python
import sys

def flush():
    sys.stdout.flush()

def ask(name):
    """특정 사람의 생일이 오늘인지 질문"""
    print(f"? {name}", flush=True)
    response = int(input().strip())
    return response

def answer(name):
    """생일인 사람을 찾았을 때 답 출력"""
    print(f"! {name}", flush=True)

def solve():
    # 입력 받기
    n = int(input().strip())
    people = []
    for _ in range(n):
        people.append(input().strip())

    # 각 사람에 대한 응답 기록
    responses = {}

    # 모든 사람에게 2번씩 질문
    for person in people:
        responses[person] = []

        # 첫 번째 질문
        response1 = ask(person)
        responses[person].append(response1)

        # 두 번째 질문
        response2 = ask(person)
        responses[person].append(response2)

        # 두 번 모두 1이면 확실히 생일인 사람
        if response1 == 1 and response2 == 1:
            answer(person)
            return

    # 두 번 모두 1인 사람이 없는 경우
    # 1이 한 번이라도 나온 사람 찾기
    candidates = []
    for person, resps in responses.items():
        if 1 in resps:
            candidates.append(person)

    # 후보가 정확히 1명이면 그 사람이 답
    if len(candidates) == 1:
        answer(candidates[0])
    else:
        # 이론적으로 여기 도달하면 안 되지만,
        # 혹시 모르니 첫 번째 후보 선택
        answer(candidates[0])

if __name__ == "__main__":
    solve()
```
