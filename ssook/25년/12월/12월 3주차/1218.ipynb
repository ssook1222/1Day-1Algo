## BoJ

### 2711

```python
def solve():
    t = int(input())

    for _ in range(t):
        parts = input().split()
        position = int(parts[0])  # 1-based index
        text = parts[1]

        # position번째 문자 제거 (0-based로 변환하면 position-1)
        # 방법 1: 슬라이싱
        result = text[:position-1] + text[position:]

        print(result)

solve()
```
