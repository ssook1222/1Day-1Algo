```python
def solve():
    n = int(input())

    # 7마리 소 이름
    cows = ["Bessie", "Elsie", "Daisy", "Gertie", "Annabelle", "Maggie", "Henrietta"]

    # 각 소의 총 우유 생산량 초기화
    milk = {cow: 0 for cow in cows}

    # 로그 읽기
    for _ in range(n):
        parts = input().split()
        cow_name = parts[0]
        amount = int(parts[1])
        milk[cow_name] += amount

    # 생산량 리스트로 변환
    productions = list(milk.values())

    # 최소값 찾기
    min_production = min(productions)

    # 최소값보다 큰 값들만 추출
    greater_than_min = [p for p in productions if p > min_production]

    # 최소값보다 큰 값이 없으면 모두 동일
    if not greater_than_min:
        print("Tie")
        return

    # 두 번째로 작은 값 (최소값보다 큰 값 중 최소)
    second_min = min(greater_than_min)

    # 두 번째로 작은 값을 가진 소 찾기
    second_cows = [cow for cow, prod in milk.items() if prod == second_min]

    # 여러 소가 동점이면 Tie
    if len(second_cows) > 1:
        print("Tie")
    else:
        print(second_cows[0])

solve()
```
