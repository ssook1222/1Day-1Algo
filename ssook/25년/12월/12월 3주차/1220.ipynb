## BoJ

### 18406

```python
def solve():
    n = input().strip()

    # 자릿수의 절반
    length = len(n)
    half = length // 2

    # 왼쪽 부분과 오른쪽 부분으로 나누기
    left_part = n[:half]
    right_part = n[half:]

    # 각 부분의 자릿수 합 계산
    left_sum = sum(int(digit) for digit in left_part)
    right_sum = sum(int(digit) for digit in right_part)

    # 두 합이 같으면 LUCKY, 다르면 READY
    if left_sum == right_sum:
        print("LUCKY")
    else:
        print("READY")

solve()
```
