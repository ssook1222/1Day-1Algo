### 전화번호 목록

```python
def solution(phone_book):
    # 해시 셋에 모든 전화번호 저장
    phone_set = set(phone_book)

    # 각 전화번호의 모든 접두어를 확인
    for phone in phone_book:
        # 전화번호의 모든 접두어를 확인
        for i in range(1, len(phone)):
            prefix = phone[:i]
            if prefix in phone_set:
                return False

    return True
```

```python
# 입력
n = int(input())  # 스위치 개수
switches = list(map(int, input().split()))  # 스위치 상태 (0-indexed로 사용)
students = int(input())  # 학생 수

for _ in range(students):
    gender, num = map(int, input().split())

    if gender == 1:  # 남학생
        # 받은 수의 배수 번호 스위치 상태 변경
        for i in range(num - 1, n, num):  # num의 배수들 (0-indexed)
            switches[i] = 1 - switches[i]  # 0->1, 1->0

    else:  # 여학생 (gender == 2)
        # 받은 번호를 중심으로 좌우 대칭 구간 찾기
        center = num - 1  # 0-indexed
        switches[center] = 1 - switches[center]  # 중심은 항상 바뀜

        # 좌우로 확장하면서 대칭 확인
        left = center - 1
        right = center + 1

        while left >= 0 and right < n and switches[left] == switches[right]:
            switches[left] = 1 - switches[left]
            switches[right] = 1 - switches[right]
            left -= 1
            right += 1

# 출력 (한 줄에 20개씩)
for i in range(n):
    print(switches[i], end='')
    if (i + 1) % 20 == 0:  # 20개마다 줄바꿈
        print()
    elif i < n - 1:  # 마지막이 아니면 공백
        print(' ', end='')
    else:  # 마지막
        print()
```
