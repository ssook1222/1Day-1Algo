## BoJ

```python
import math

def solve():
    n, k = map(int, input().split())

    # (성별, 학년)별 학생 수 카운트
    # students[성별][학년] = 학생 수
    students = [[0] * 7 for _ in range(2)]  # 0: 여학생, 1: 남학생, 학년: 1~6

    for _ in range(n):
        gender, grade = map(int, input().split())
        students[gender][grade] += 1

    # 필요한 방의 총 개수 계산
    total_rooms = 0

    for gender in range(2):  # 0: 여학생, 1: 남학생
        for grade in range(1, 7):  # 1학년 ~ 6학년
            student_count = students[gender][grade]
            if student_count > 0:
                # 올림 계산: (학생 수 + K - 1) // K 또는 math.ceil(학생 수 / K)
                rooms_needed = math.ceil(student_count / k)
                total_rooms += rooms_needed

    print(total_rooms)

solve()
```
