```python
from collections import Counter

# 입력 받기
n = int(input())

# 모든 태그를 저장할 리스트
all_tags = []

for _ in range(n):
    line = input().split()
    s = int(line[0])  # 문제 번호
    t = int(line[1])  # 태그 개수
    tags = line[2:2+t]  # 태그 목록

    all_tags.extend(tags)

# 태그별 등장 횟수 계산
tag_count = Counter(all_tags)

# 가장 많이 등장한 횟수 찾기
max_count = max(tag_count.values())

# 가장 많이 등장한 태그들 찾기
most_common_tags = [tag for tag, count in tag_count.items() if count == max_count]

# 결과 출력
if len(most_common_tags) == 1:
    print(most_common_tags[0])
else:
    print(-1)
```
