```python
def check_win(board, player):
    """해당 플레이어가 이기는 위치를 찾는 함수"""
    # 모든 빈 칸에 대해 확인
    for i in range(3):
        for j in range(3):
            if board[i][j] == '-':
                # 임시로 말을 놓아봄
                board[i][j] = player

                # 승리 조건 확인
                if is_winner(board, player):
                    return i, j

                # 원상복구
                board[i][j] = '-'

    return -1, -1

def is_winner(board, player):
    """해당 플레이어가 승리했는지 확인"""
    # 가로 확인
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] == player:
            return True

    # 세로 확인
    for j in range(3):
        if board[0][j] == board[1][j] == board[2][j] == player:
            return True

    # 대각선 확인
    if board[0][0] == board[1][1] == board[2][2] == player:
        return True
    if board[0][2] == board[1][1] == board[2][0] == player:
        return True

    return False

# 입력 받기
t = int(input())

for case_num in range(1, t + 1):
    # 게임판 입력
    board = []
    for _ in range(3):
        board.append(list(input().strip()))

    # 남규의 말
    player = input().strip()

    # 승리 위치 찾기
    win_i, win_j = check_win(board, player)

    # 말 놓기
    board[win_i][win_j] = player

    # 출력
    print(f"Case {case_num}:")
    for row in board:
        print(''.join(row))
```
