```python
# 입력 받기
w, n = input().split()
n = int(n)

# 배열 입력
grid = []
for _ in range(n):
    row = list(map(int, input().split()))
    grid.append(row)

# 뒤집기 규칙: 2↔5, 1→1, 8→8, 나머지는 ?
def flip_number(num):
    if num == 1:
        return 1
    elif num == 2:
        return 5
    elif num == 5:
        return 2
    elif num == 8:
        return 8
    else:
        return '?'

# 방향에 따라 뒤집기
if w == 'U':  # 상하 뒤집기 (위아래 반전)
    # 행의 순서를 뒤집고, 각 숫자도 뒤집기
    result = []
    for i in range(n-1, -1, -1):
        row = [flip_number(grid[i][j]) for j in range(n)]
        result.append(row)

elif w == 'D':  # 하상 뒤집기 (위아래 반전) - U와 동일
    result = []
    for i in range(n-1, -1, -1):
        row = [flip_number(grid[i][j]) for j in range(n)]
        result.append(row)

elif w == 'L':  # 좌우 뒤집기 (좌우 반전)
    result = []
    for i in range(n):
        row = [flip_number(grid[i][j]) for j in range(n-1, -1, -1)]
        result.append(row)

elif w == 'R':  # 우좌 뒤집기 (좌우 반전) - L과 동일
    result = []
    for i in range(n):
        row = [flip_number(grid[i][j]) for j in range(n-1, -1, -1)]
        result.append(row)

# 출력
for row in result:
    print(' '.join(map(str, row)))
```
