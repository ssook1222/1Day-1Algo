```python
# 입력 받기
d = float(input())  # 테이블의 지름
w = float(input())  # 각 기사가 필요한 공간
n = int(input())    # 기사의 수

# 원주율
pi = 3.14159

# 테이블의 둘레 계산
circumference = pi * d

# 필요한 총 공간 계산
required_space = w * n

# 판단
if circumference >= required_space:
    print("YES")
else:
    print("NO")
```
