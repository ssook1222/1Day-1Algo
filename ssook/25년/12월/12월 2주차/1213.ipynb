```python
def count_nines(s):
    """문자열에서 '9'의 개수를 센다"""
    return str(s).count('9')

# 입력
n, m = map(int, input().split())
board = []
for _ in range(n):
    row = list(map(int, input().split()))
    board.append(row)

# 전체 빙고판에서 9의 개수 세기
total_nines = 0
for i in range(n):
    for j in range(m):
        total_nines += count_nines(board[i][j])

# 각 행에서 9의 개수
row_nines = []
for i in range(n):
    count = 0
    for j in range(m):
        count += count_nines(board[i][j])
    row_nines.append(count)

# 각 열에서 9의 개수
col_nines = []
for j in range(m):
    count = 0
    for i in range(n):
        count += count_nines(board[i][j])
    col_nines.append(count)

# 가장 많은 9를 가진 행 또는 열의 9 개수
max_nines_to_remove = max(max(row_nines), max(col_nines))

# 남은 9의 개수
remaining_nines = total_nines - max_nines_to_remove

print(remaining_nines)
```
