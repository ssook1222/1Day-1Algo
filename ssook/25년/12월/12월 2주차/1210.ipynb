```python
# 입력 받기
a, b = map(int, input().split())
R = int(input())

# 각 시간에 꽃이 피는 좌표를 저장
positions = {}
x, y = a, b
t = 0

# 첫 번째 꽃 (트릭 플라워)
positions[(x, y)] = t

while True:
    t += 1

    # 다음 좌표 계산
    if (x + 1) + (y + 1) < R:
        x = x + 1
        y = y + 1
    else:
        x = x // 2
        y = y // 2

    # 이미 해당 좌표에 꽃이 있는지 확인
    if (x, y) in positions:
        print(t)
        break

    # 새로운 좌표 저장
    positions[(x, y)] = t
```
