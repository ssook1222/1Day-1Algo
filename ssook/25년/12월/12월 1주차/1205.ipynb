```python
def count_steps_back(heights):
    """
    학생들이 줄을 설 때 뒤로 물러서는 총 횟수를 계산

    Args:
        heights: 학생들의 키 리스트 (줄서기 순서대로)

    Returns:
        총 뒤로 물러선 횟수
    """
    line = []  # 현재 줄에 서 있는 학생들의 키
    total_steps = 0

    for height in heights:
        # 현재 학생이 들어갈 위치 찾기
        # 자기보다 키가 큰 학생 중 가장 앞에 있는 학생의 위치
        insert_pos = len(line)  # 기본값: 맨 뒤

        for i in range(len(line)):
            if line[i] > height:
                insert_pos = i
                break

        # 삽입 위치부터 끝까지의 학생들이 한 칸씩 뒤로 물러남
        steps = len(line) - insert_pos
        total_steps += steps

        # 학생을 해당 위치에 삽입
        line.insert(insert_pos, height)

    return total_steps


# 테스트 케이스 개수 입력
P = int(input())

# 각 테스트 케이스 처리
for _ in range(P):
    data = list(map(int, input().split()))
    test_case_num = data[0]
    heights = data[1:21]

    # 뒤로 물러선 총 횟수 계산
    result = count_steps_back(heights)

    # 결과 출력
    print(f"{test_case_num} {result}")
```
