## BoJ

### 6766

```python
K = int(input())
encoded_word = input()

decoded_word = ""

for P in range(1, len(encoded_word) + 1):
    # 현재 문자
    char = encoded_word[P - 1]

    # 시프트 값 계산
    S = 3 * P + K

    # 문자를 숫자로 변환 (A=0, B=1, ..., Z=25)
    char_num = ord(char) - ord('A')

    # 디코딩: S만큼 뒤로 이동
    original_num = (char_num - S) % 26

    # 숫자를 문자로 변환
    original_char = chr(original_num + ord('A'))

    decoded_word += original_char

print(decoded_word)
```
