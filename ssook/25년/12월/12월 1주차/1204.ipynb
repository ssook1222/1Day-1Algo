```python
def find_longest_x_subsequence(sequence):
    """연속된 X의 최대 길이를 찾는 함수"""
    max_length = 0
    current_length = 0

    for item in sequence:
        if item == 'X':
            current_length += 1
            max_length = max(max_length, current_length)
        else:
            current_length = 0

    return max_length


# 테스트 케이스 개수 입력
t = int(input())

# 각 테스트 케이스 처리
for _ in range(t):
    line = input().split()
    n = int(line[0])
    sequence = line[1:n+1]

    # 최대 연속 X 길이 찾기
    max_length = find_longest_x_subsequence(sequence)

    # 결과 출력
    print(f"The longest contiguous subsequence of X's is of length {max_length}")
```
