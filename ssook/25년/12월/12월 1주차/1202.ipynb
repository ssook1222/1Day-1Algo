## BoJ

### 14409

```python
N = int(input())
X = int(input())

total_value = 0

for _ in range(N):
    line = input().split()
    P1 = int(line[0])
    P2 = int(line[1])

    # P1과 P2의 차이 계산
    diff = abs(P1 - P2)

    if diff <= X:
        # 차이가 X 이하면 더 큰 값을 선택
        price = max(P1, P2)
    else:
        # 차이가 X보다 크면 P3를 읽어서 사용
        P3 = int(input())
        price = P3

    total_value += price

print(total_value)
```
