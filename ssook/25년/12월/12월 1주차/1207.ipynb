```python
def matches_pattern(scanned, plate):
    """
    스캔된 코드와 차량 번호판이 매칭되는지 확인

    Args:
        scanned: 스캔된 코드 (9자리, *는 와일드카드)
        plate: 데이터베이스의 차량 번호판 (9자리)

    Returns:
        매칭 여부 (True/False)
    """
    # 길이가 9가 아니면 매칭 안됨
    if len(scanned) != 9 or len(plate) != 9:
        return False

    # 각 위치별로 비교
    for i in range(9):
        # *는 모든 문자와 매칭
        if scanned[i] == '*':
            continue
        # *가 아닌 경우 정확히 일치해야 함
        if scanned[i] != plate[i]:
            return False

    return True


# 스캔된 코드 입력
scanned_code = input().strip()

# 데이터베이스 코드 개수 입력
n = int(input())

# 매칭되는 코드 찾기
matching_codes = []
for _ in range(n):
    plate = input().strip()
    if matches_pattern(scanned_code, plate):
        matching_codes.append(plate)

# 결과 출력
print(len(matching_codes))
for code in matching_codes:
    print(code)
```
