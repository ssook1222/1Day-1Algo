```python
def find_triangle(segments):
    """
    주어진 선분들 중 삼각형을 만들 수 있는 3개를 찾기

    삼각형 조건: 세 변 a ≤ b ≤ c에 대해 a + b > c

    Args:
        segments: 선분 길이 리스트

    Returns:
        삼각형을 만들 수 있는 3개의 변 또는 None
    """
    # 적어도 3개가 필요
    if len(segments) < 3:
        return None

    # 내림차순 정렬 (큰 것부터 확인)
    segments.sort(reverse=True)

    # 인접한 3개씩 확인
    for i in range(len(segments) - 2):
        a = segments[i]
        b = segments[i + 1]
        c = segments[i + 2]

        # a >= b >= c 이므로 a + b > c만 확인하면 됨
        # (다른 조건들은 자동으로 만족)
        if b + c > a:
            return (a, b, c)

    return None


# 입력 받기
segments = []
while True:
    n = int(input())
    if n == 0:
        break
    segments.append(n)

# 삼각형 찾기
result = find_triangle(segments)

# 결과 출력
if result:
    print(f"{result[0]} {result[1]} {result[2]}")
else:
    print("NIE")
```
