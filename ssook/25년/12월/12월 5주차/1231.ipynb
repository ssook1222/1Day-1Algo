## BoJ

```python
# 입력 받기
C = int(input())
M = int(input())
primes = list(map(int, input().split()))

# 빠른 검색을 위해 set으로 변환
prime_set = set(primes)

# 각 소수로 C를 나누어보기
for p in primes:
    if C % p == 0:  # p로 나누어떨어지면
        other = C // p  # 다른 한 변의 길이

        # other가 리스트에 있고, p와 다른지 확인
        if other in prime_set and other != p:
            # 작은 값을 먼저 출력
            result = sorted([p, other])
            print(result[0], result[1])
            break
```
