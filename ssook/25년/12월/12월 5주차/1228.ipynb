## BoJ

### 30034

```python
N = int(input())
char_separators = set(input().split())

M = int(input())
num_separators = set(input().split())

K = int(input())
mergers = set(input().split())

S = int(input())
text = input()

# 병합자로 주어진 것들은 구분자에서 제외
char_separators -= mergers
num_separators -= mergers

# 모든 구분자를 합침
all_separators = char_separators | num_separators

# 문자열을 나누기
result = []
current = ""

for char in text:
    if char == ' ' or char in all_separators:
        # 구분자를 만났을 때
        if current:  # 현재까지 모은 문자열이 있으면 결과에 추가
            result.append(current)
            current = ""
    else:
        # 일반 문자는 현재 문자열에 추가
        current += char

# 마지막 남은 문자열 처리
if current:
    result.append(current)

# 결과 출력
for word in result:
    print(word)
```
