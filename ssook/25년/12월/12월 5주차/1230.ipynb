## BoJ

### 15780

```python
# 입력 받기
N, K = map(int, input().split())
multitaps = list(map(int, input().split()))

# 각 멀티탭에서 사용 가능한 구의 개수 계산
# n구 멀티탭에서 연속되지 않게 꽂을 수 있는 최대 개수 = (n + 1) // 2
total_available = 0
for multitap in multitaps:
    available = (multitap + 1) // 2
    total_available += available

# N명이 모두 사용할 수 있는지 확인
if total_available >= N:
    print("YES")
else:
    print("NO")
```
