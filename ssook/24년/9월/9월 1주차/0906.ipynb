```python
import sys
sys.setrecursionlimit(10000)

def dfs(x, y):
    if y == end_col:  # 마지막 열에 도달했을 경우
        return True

    for i in range(3):
        nx = x + dx[i]
        ny = y + 1
        if 0 <= nx < rows and 0 <= ny < cols and graph[nx][ny] == '.':
            graph[nx][ny] = 'x'
            if dfs(nx, ny):
                return True
    return False

rows, cols = map(int, input().split())
graph = []
for _ in range(rows):
    graph.append(list(input().strip()))  # 각 행을 리스트로 변환하여 그래프에 추가

dx = [-1, 0, 1]  # 이동 방향 (오른쪽, 오른쪽 위 대각선, 오른쪽 아래 대각선)
end_col = cols - 1  # 마지막 열의 인덱스
max_pipes = 0

# 모든 가능한 경로 탐색
for x in range(rows):
    if dfs(x, 0):
        max_pipes += 1

print(max_pipes)
```
