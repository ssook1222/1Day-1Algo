### 2437

```python
import sys
input = sys.stdin.readline

n = int(input().strip())
arr = list(map(int, input().split()))
arr.sort()

target = 1

for num in arr:
    if target < num:
        break

    target += num

print(target)
```

### 작은 수에서 빼기


```python
import sys

input = sys.stdin.read
data = input().strip().split()

a = int(data[0])
b = int(data[1])

result = a - b if a < b else b - a

print(result)
```
