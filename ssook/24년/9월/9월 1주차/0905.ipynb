### 10775

```python
import sys

input = sys.stdin.read
data = input().strip().split()

num_gates = int(data[0])
num_airplanes = int(data[1])
airplanes = [int(data[i + 2]) for i in range(num_airplanes)]

alters = list(range(num_gates + 1))  # 대안 게이트

def find_root(airplane):
    stack = [airplane]
    while alters[airplane] != airplane:
        stack.append(alters[airplane])
        airplane = alters[airplane]
    while stack:
        alters[stack.pop()] = alters[airplane]
    return airplane

def union(a, b):
    a_root = find_root(a)
    b_root = find_root(b)
    alters[a_root] = b_root

cnt = 0

for airplane in airplanes:
    root = find_root(airplane)
    if root == 0:
        break
    union(root, root - 1)
    cnt += 1

print(cnt)
```
