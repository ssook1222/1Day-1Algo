```python
import sys
from copy import deepcopy
from collections import deque
from itertools import combinations

def castle_defense(archers):
    board_copy = deepcopy(board)  # 보드 복사
    visited = [[0] * M for _ in range(N)]  # 방문 처리
    count = 0

    for i in range(N-1, -1, -1):  # 한 칸씩 전진 (적이 아닌 궁수가 전진)
        to_be_removed = []
        for archer_col in archers:  # 뽑은 궁수 자리에 대해
            queue = deque([[i, archer_col, 1]])
            while queue:
                y, x, d = queue.popleft()
                if board_copy[y][x] == 1:
                    to_be_removed.append([y, x])
                    if not visited[y][x]:
                        visited[y][x] = 1
                        count += 1  # 적을 제거하고 점수 증가
                    break

                if d < D:  # 사정거리보다 짧으면
                    for dx, dy in move:
                        nx, ny = x + dx, y + dy
                        if 0 <= ny < N and 0 <= nx < M:
                            queue.append([ny, nx, d + 1])  # 거리 증가

        for y, x in to_be_removed:  # 모든 궁수가 동시에 공격하니 마지막에 처리
            board_copy[y][x] = 0

    return count

# 입력 받기
N, M, D = map(int, sys.stdin.readline().split())
board = [list(map(int, sys.stdin.readline().split())) for _ in range(N)]  # 격자판

move = [[-1, 0], [0, -1], [1, 0]]  # 이동할 수 있는 방향

result = 0
# 궁수 배치하기
for archers in combinations([i for i in range(M)], 3):  # 궁수 3명의 모든 조합
    result = max(result, castle_defense(archers))

print(result)
```

### 11437

```python
import sys
sys.setrecursionlimit(100000)

n = int(input())
parent = [0] * (n + 1)      # 각 노드의 부모 노드 정보
depth = [0] * (n + 1)       # 각 노드까지의 깊이
visited = [False] * (n + 1)  # 방문 여부
graph = [[] for _ in range(n + 1)]

for _ in range(n - 1):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

# 루트 노드부터의 깊이 구하기
def dfs(node, d):
    visited[node] = True
    depth[node] = d

    for neighbor in graph[node]:
        if not visited[neighbor]:
            parent[neighbor] = node
            dfs(neighbor, d + 1)

# 최소 공통 조상 찾기
def lca(u, v):
    # 깊이 맞추기
    while depth[u] != depth[v]:
        if depth[u] > depth[v]:
            u = parent[u]
        else:
            v = parent[v]

    # 노드 맞추기
    while u != v:
        u = parent[u]
        v = parent[v]

    return u

dfs(1, 0)

m = int(input())

for _ in range(m):
    u, v = map(int, input().split())
    print(lca(u, v))
```
