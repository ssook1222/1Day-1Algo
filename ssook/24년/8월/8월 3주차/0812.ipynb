### 11779

```python
import sys
from collections import defaultdict
import heapq

INF = int(1e9)

num_cities = int(sys.stdin.readline().rstrip())
num_buses = int(sys.stdin.readline().rstrip())
graph = defaultdict(list)

for _ in range(num_buses):
    departure, destination, cost = map(int, sys.stdin.readline().rstrip().split())
    graph[departure].append((destination, cost))

start_city, end_city = map(int, sys.stdin.readline().rstrip().split())

dist = [INF] * (num_cities + 1)
prev_node = [0] * (num_cities + 1)

def dijkstra(start):
    pq = []
    heapq.heappush(pq, (0, start))
    dist[start] = 0

    while pq:
        current_dist, current_node = heapq.heappop(pq)

        if dist[current_node] < current_dist:
            continue

        for neighbor, weight in graph[current_node]:
            distance = current_dist + weight

            if distance < dist[neighbor]:
                dist[neighbor] = distance
                prev_node[neighbor] = current_node
                heapq.heappush(pq, (distance, neighbor))

dijkstra(start_city)

print(dist[end_city])

path = []
current = end_city
while current != 0:
    path.append(current)
    current = prev_node[current]
path.reverse()

print(len(path))
print(' '.join(map(str, path)))
```

### 17471

```python
import sys
from itertools import combinations
from collections import deque

# 입력 받기
num_areas = int(sys.stdin.readline().rstrip())
population = list(map(int, sys.stdin.readline().rstrip().split()))
adjacency_list = [[0] * num_areas for _ in range(num_areas)]

for i in range(num_areas):
    data = list(map(int, sys.stdin.readline().rstrip().split()))
    num_neighbors = data[0]
    for j in range(1, num_neighbors + 1):
        adjacency_list[i][data[j] - 1] = 1  # 구역 번호를 0부터 시작하도록 조정

def is_connected(nodes):
    q = deque()
    check = [False] * num_areas
    q.append(nodes[0])
    check[nodes[0]] = True

    while q:
        node = q.popleft()

        for i in range(num_areas):
            if adjacency_list[node][i] == 0:
                continue
            if i not in nodes:
                continue
            if not check[i]:
                check[i] = True
                q.append(i)

    return check.count(True) == len(nodes)

def get_total_population(nodes):
    total = 0
    for node in nodes:
        total += population[node]
    return total

# 가능한 선거구 나누기 찾기
INF = int(1e9)
min_population_diff = INF
found_valid_division = False
all_areas = set(range(num_areas))

for i in range(1, num_areas // 2 + 1):
    for A in combinations(all_areas, i):
        B = list(all_areas - set(A))

        if is_connected(A) and is_connected(B):
            population_diff = abs(get_total_population(A) - get_total_population(B))
            min_population_diff = min(min_population_diff, population_diff)
            found_valid_division = True

# 결과 출력
if found_valid_division:
    print(min_population_diff)
else:
    print(-1)
```
