```python
from collections import deque

# 입력 받기
n, m = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n)]

# 방문 여부를 나타내는 visited 배열
visited = [[0]*m for _ in range(n)]

# 상하좌우 이동을 위한 배열
dxs, dys = [0, 0, 1, -1], [1, -1, 0, 0]

# 각 모양의 크기를 저장할 사전
shape_sizes = dict()

# 범위 체크 함수
def in_range(x, y):
    return 0 <= x < n and 0 <= y < m

# BFS 함수 정의
def bfs(x, y, cnt):
    visited[x][y] = cnt
    que = deque([(x, y)])
    while que:
        dx, dy = que.popleft()
        for i in range(4):
            nx, ny = dx + dxs[i], dy + dys[i]
            if in_range(nx, ny) and not visited[nx][ny] and graph[nx][ny] == 1:
                visited[nx][ny] = cnt
                que.append((nx, ny))

# 영역 구별하기
cnt = 0
for i in range(n):
    for j in range(m):
        if graph[i][j] == 1 and not visited[i][j]:
            cnt += 1
            bfs(i, j, cnt)

# 각 영역의 크기 계산하여 저장하기
for i in range(n):
    for j in range(m):
        if visited[i][j] > 0:
            if visited[i][j] not in shape_sizes:
                shape_sizes[visited[i][j]] = 1
            else:
                shape_sizes[visited[i][j]] += 1

# 그래프의 0인 부분을 기준으로 상하좌우를 체크하며 영역의 최대 크기 구하기
answer = 0
for i in range(n):
    for j in range(m):
        if graph[i][j] == 0:
            adjacent_shapes = set()
            for t in range(4):
                nx, ny = i + dxs[t], j + dys[t]
                if in_range(nx, ny) and visited[nx][ny]:
                    adjacent_shapes.add(visited[nx][ny])
            result = 0
            for k in adjacent_shapes:
                result += shape_sizes[k]
            answer = max(answer, result + 1)

print(answer)
```

```python
passengers = []
num_houses, goal = map(int, input().split())

# 입력을 받고 출발지와 목적지를 기준으로 정리
for _ in range(num_houses):
    start, end = map(int, input().split())
    if start > end:
        passengers.append([end, start])  # 시작점이 더 큰 경우 순서를 바꿔서 저장
    else:
        passengers.append([start, end])

# 목적지를 기준으로 정렬 (목적지가 큰 순서대로 정렬)
passengers.sort(key=lambda x: -x[1])

merged_trips = []
start, end = passengers[0]

# 겹치는 구간을 합쳐서 merged_trips에 저장
for i in range(1, len(passengers)):
    s, e = passengers[i]
    if start <= e:  # 현재 구간과 다음 구간이 겹치는 경우
        start = min(start, s)
    else:  # 겹치지 않는 경우
        merged_trips.append([start, end])
        start, end = s, e

merged_trips.append([start, end])

total_distance = goal
# 합쳐진 구간들 간의 이동 거리를 계산하여 최소 거리를 구함
for i in range(len(merged_trips)):
    total_distance += 2 * (merged_trips[i][1] - merged_trips[i][0])

print(total_distance)
```

### 코드네임

```python
class Agent:
    def __init__(self, code_name, score):
        self.code_name = code_name
        self.score = int(score)  # 입력된 점수는 문자열로 변환하여 정수로 변환하여 저장

# 5명의 요원 객체 생성
agents = []
for _ in range(5):
    data = input().split()
    code_name = data[0]
    score = data[1]
    agents.append(Agent(code_name, score))

# 가장 점수가 낮은 요원 찾기
min_score = float('inf')
min_agent = None
for agent in agents:
    if agent.score < min_score:
        min_score = agent.score
        min_agent = agent

# 결과 출력
print(f"{min_agent.code_name} {min_agent.score}")
```
