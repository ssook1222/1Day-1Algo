```python
import sys

sys.setrecursionlimit(10 ** 8)
input = lambda: sys.stdin.readline().rstrip()


def catch_shark(col):
    for row in range(rows):
        if grid[row][col]:
            size = grid[row][col][2]
            grid[row][col] = 0
            return size
    return 0


def move_sharks():
    global grid  # grid[row][col]에는 (speed, direction, size)의 값이 들어 있음. 상어가 없는 경우엔 0이 들어 있음
    new_grid = [[0 for _ in range(cols)] for _ in range(rows)]  # 상어들의 새 위치를 담을 공간
    for row in range(rows):
        for col in range(cols):
            if grid[row][col]:
                # 이 상어의 다음 위치는
                new_row, new_col, new_dir = compute_next_location(row, col, grid[row][col][0], grid[row][col][1])
                if new_grid[new_row][new_col]:
                    new_grid[new_row][new_col] = max(
                        new_grid[new_row][new_col],
                        (grid[row][col][0], new_dir, grid[row][col][2]),
                        key=lambda x: x[2],
                    )
                else:
                    new_grid[new_row][new_col] = (grid[row][col][0], new_dir, grid[row][col][2])

    grid = new_grid  # grid가 이제 상어들의 새 위치를 가리키도록


def compute_next_location(row, col, speed, direction):
    if direction == UP or direction == DOWN:  # i 방향
        cycle = rows * 2 - 2
        if direction == UP:
            speed += 2 * (rows - 1) - row
        else:
            speed += row

        speed %= cycle
        if speed >= rows:
            return (2 * rows - 2 - speed, col, UP)
        return (speed, col, DOWN)

    else:  # j 방향
        cycle = cols * 2 - 2
        if direction == LEFT:
            speed += 2 * (cols - 1) - col
        else:
            speed += col

        speed %= cycle
        if speed >= cols:
            return (row, 2 * cols - 2 - speed, LEFT)
        return (row, speed, RIGHT)


UP, DOWN, RIGHT, LEFT = 1, 2, 3, 4
rows, cols, num_sharks = map(int, input().split())
grid = [[0 for _ in range(cols)] for _ in range(rows)]

for _ in range(num_sharks):
    r, c, s, d, z = map(int, input().split())
    r, c = r - 1, c - 1
    grid[r][c] = (s, d, z)
    # s : speed
    # d : 1(up), 2(down), 3(right), 4(left)
    # z : size


total_size = 0
for col in range(cols):
    total_size += catch_shark(col)
    move_sharks()

print(total_size)
```

### 정렬된 숫자 위치 알아내기

```python
# 입력을 읽기 위한 모듈
import sys
input = sys.stdin.readline

# 수열의 길이와 원소를 입력받는다.
N = int(input().strip())
sequence = list(map(int, input().strip().split()))

# 원래의 인덱스를 기억하기 위해 (값, 인덱스) 형태로 저장한다.
indexed_sequence = [(sequence[i], i) for i in range(N)]

# 원소를 오름차순으로 정렬한다.
sorted_sequence = sorted(indexed_sequence)

# 결과를 저장할 리스트 (원래 인덱스에 대한 이동 위치)
result = [0] * N

# 정렬된 순서에 따라 결과를 업데이트한다.
for new_index, (value, old_index) in enumerate(sorted_sequence):
    result[old_index] = new_index + 1  # 1부터 시작하는 위치를 기록

# 결과를 출력한다.
print(" ".join(map(str, result)))
```
