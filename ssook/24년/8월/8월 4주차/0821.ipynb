```python
n = int(input())
sequence = list(map(int, input().split()))

def find_position(start, end, target):
    if start > end:
        return start

    mid = (start + end) // 2

    if lis[mid] > target:
        return find_position(start, mid - 1, target)
    elif lis[mid] == target:
        return mid
    else:
        return find_position(mid + 1, end, target)

lis = [sequence[0]]

for i in range(1, len(sequence)):
    if lis[-1] < sequence[i]:
        lis.append(sequence[i])
    else:
        pos = find_position(0, len(lis) - 1, sequence[i])
        lis[pos] = sequence[i]

print(len(lis))
```

### 키, 몸무게를 기준으로 정렬

```python
# 입력 받기
n = int(input())

# 사람들의 정보를 저장할 리스트
people = []

# 각 사람의 정보 입력 받기
for _ in range(n):
    data = input().split()
    name = data[0]
    height = int(data[1])
    weight = int(data[2])
    people.append((name, height, weight))

# 정렬 함수
# 키를 기준으로 오름차순 정렬, 키가 동일한 경우 몸무게를 기준으로 내림차순 정렬
people_sorted = sorted(people, key=lambda person: (person[1], -person[2]))

# 결과 출력
for person in people_sorted:
    print(person[0], person[1], person[2])
```
