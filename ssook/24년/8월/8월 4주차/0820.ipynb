```python
from collections import defaultdict

def fibonacci(n):
    if n <= 1:
        return memo[n]
    # 중복 연산 막음 : 메모 활용
    elif memo[n] > 0:
        return memo[n]
    # 해시값 존재 안할 시 최소 한번은 연산해주어야 함.
    else:
        half = n // 2
        if n % 2 == 0:
            # Even case
            f_half = fibonacci(half)
            f_half_minus_one = fibonacci(half - 1)
            memo[n] = ((2 * f_half_minus_one + f_half) * f_half) % 1000000007
        else:
            # Odd case
            f_half_plus_one = fibonacci(half + 1)
            f_half = fibonacci(half)
            memo[n] = (f_half_plus_one**2 + f_half**2) % 1000000007
        return memo[n]

# 사용자로부터 n 값을 입력받는다
n = int(input())

# 피보나치 수를 메모이제이션하기 위한 딕셔너리 초기화
memo = defaultdict(int)
memo[0], memo[1] = 0, 1

# n번째 피보나치 수를 출력
print(fibonacci(n))
```

### 비오는 날

```python
from datetime import datetime

# 날짜를 비교하여 가장 빠른 비 오는 날을 찾는 함수
def find_earliest_rain_day(n, weather_data):
    # 비 오는 날의 정보를 저장할 변수 초기화
    earliest_rain_date = None
    earliest_rain_info = None

    # 모든 날씨 정보를 확인
    for data in weather_data:
        date_str, day_of_week, weather = data

        # 날씨가 'Rain'인 경우에만 처리
        if weather == 'Rain':
            # 날짜 문자열을 datetime 객체로 변환
            date = datetime.strptime(date_str, '%Y-%m-%d')

            # 첫 번째 비 오는 날 또는 현재 가장 빠른 비 오는 날보다 더 이른 날일 경우 업데이트
            if earliest_rain_date is None or date < earliest_rain_date:
                earliest_rain_date = date
                earliest_rain_info = data

    return earliest_rain_info

# 입력 받기
n = int(input())
weather_data = [input().strip().split() for _ in range(n)]

# 가장 근 시일내에 비 오는 날 찾기
earliest_rain_day = find_earliest_rain_day(n, weather_data)

# 결과 출력
print(' '.join(earliest_rain_day))
```
