## BoJ

### 1300

```python
N, K = int(input()), int(input())
low, high = 1, N * N  # 배열 A에서 가능한 값의 범위로 설정

while low <= high:
    mid_value = (low + high) // 2

    count = 0
    for i in range(1, N + 1):
        count += min(mid_value // i, N)  # i의 배수 또는 최대 N까지의 수

    if count >= K:  # 이분탐색 조건
        answer = mid_value
        high = mid_value - 1
    else:
        low = mid_value + 1

print(answer)
```

## CodeTree

### 원점으로부터의 거리

```python
import sys
input = sys.stdin.readline

# 점의 수를 입력받는다.
N = int(input().strip())

# 점들의 정보를 담을 리스트
points = []

# 각 점의 좌표와 인덱스를 입력받는다.
for i in range(N):
    x, y = map(int, input().strip().split())
    # 멘하턴 거리와 인덱스를 함께 저장한다.
    distance = abs(x) + abs(y)
    points.append((distance, i + 1))  # 인덱스는 1부터 시작하므로 i + 1

# 멘하턴 거리를 기준으로 오름차순 정렬하고, 거리 같은 경우에는 인덱스 기준으로 오름차순 정렬한다.
points.sort()

# 정렬된 점의 인덱스를 출력한다.
for _, index in points:
    print(index)
```
