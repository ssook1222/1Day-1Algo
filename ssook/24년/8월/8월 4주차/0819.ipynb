### 2749

```python
# 사용자로부터 n 값을 입력받는다
n = int(input())

# 모듈로 연산을 위한 값 정의
mod = 1000000

# 피보나치 수열의 초기 두 값 정의
fib = [0, 1]

# 피보나치 수열의 주기(p) 계산
# mod=1000000일 때 피보나치 수열의 주기(p)는 mod//10*15
period = mod // 10 * 15

# 피보나치 수열을 period까지 계산
for i in range(2, period):
    fib.append(fib[i - 1] + fib[i - 2])
    fib[i] %= mod

# n이 period로 나눈 나머지에 해당하는 피보나치 수를 출력
print(fib[n % period])
```

## CodeTree

### 코드네임

```python
class Agent:
    def __init__(self, codename, score):
        self.codename = codename
        self.score = score

# 리스트를 만들어 5명의 Agent 객체를 저장할 준비를 한다.
agents = []

# 5명의 정보를 입력받는다.
for _ in range(5):
    data = input().split()
    codename = data[0]
    score = int(data[1])
    agents.append(Agent(codename, score))

# 점수가 가장 낮은 요원을 찾는다.
lowest_agent = min(agents, key=lambda agent: agent.score)

# 결과를 출력한다.
print(lowest_agent.codename, lowest_agent.score)
```
