```python
import sys

if __name__ == "__main__":
    N, M = map(int, sys.stdin.readline().split())
    ride_times = list(map(int, sys.stdin.readline().split()))

    if N < M:
        print(N)
    else:
        # Binary search
        left, right = 0, 60000000000
        last_ride_time = None
        while left <= right:
            mid = (left + right) // 2
            cnt = M
            for i in range(M):
                cnt += mid // ride_times[i]
            if cnt >= N:
                last_ride_time = mid
                right = mid - 1
            else:
                left = mid + 1

        cnt = M
        for i in range(M):
            cnt += (last_ride_time - 1) // ride_times[i]

        for i in range(M):
            if last_ride_time % ride_times[i] == 0:
                cnt += 1
            if cnt == N:
                print(i + 1)
                break
```

### 여러가지 진수변환

```python
def decimal_to_baseN(N, B):
    result = ""
    while N > 0:
        remainder = N % B
        result = str(remainder) + result
        N //= B
    return result

# 입력 받기
import sys
input = sys.stdin.read
data = input().strip().split()
N = int(data[0])
B = int(data[1])

# B진수로 변환하여 출력
result = decimal_to_baseN(N, B)
print(result)
```
