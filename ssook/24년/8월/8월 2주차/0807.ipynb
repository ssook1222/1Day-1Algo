### 1043

```python
import sys
input = sys.stdin.readline

N, M = map(int, input().split())
truth_knowing_people = set(input().split()[1:])
parties = []

for _ in range(M):
    parties.append(set(input().split()[1:]))

for _ in range(M):
    for party in parties:
        if party & truth_knowing_people:
            truth_knowing_people = truth_knowing_people.union(party)

cnt = 0
for party in parties:
    if party & truth_knowing_people:
        continue
    cnt += 1

print(cnt)
```

### 13913

```python
from collections import deque

def print_path(x):
    arr = []
    temp = x
    for _ in range(dist[x] + 1):
        arr.append(temp)
        temp = move[temp]
    print(' '.join(map(str, arr[::-1])))

def bfs():
    q = deque()
    q.append(N)
    while q:
        x = q.popleft()
        if x == K:
            print(dist[x])
            print_path(x)
            return
        for i in (x + 1, x - 1, 2 * x):
            if 0 <= i <= 100000 and dist[i] == 0:
                q.append(i)
                dist[i] = dist[x] + 1
                move[i] = x

N, K = map(int, input().split())
dist = [0] * 100001
move = [0] * 100001
bfs()
```
