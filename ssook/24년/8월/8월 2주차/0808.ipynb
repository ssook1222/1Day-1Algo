### 2636

```python
import sys
from collections import deque

def input():
    return sys.stdin.readline().rstrip()

def bfs():
    q = deque([(0, 0)])
    melt = deque([])
    while q:
        x, y = q.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 <= nx < n and 0 <= ny < m and not visited[nx][ny]:
                visited[nx][ny] = 1
                if cheese[nx][ny] == 0:  # 공기면 계속 탐색하기 위해 큐에 넣음
                    q.append((nx, ny))
                elif cheese[nx][ny] == 1:  # 치즈면 한 번에 녹이기 위해 melt에 넣음
                    melt.append((nx, ny))

    for x, y in melt:
        cheese[x][y] = 0  # 공기와 닿은 치즈를 한 번에 녹임
    return len(melt)  # 녹인 치즈 조각의 개수 리턴

# 입력 받기
n, m = map(int, input().split())
cheese = []
total_cheese_count = 0

for i in range(n):
    cheese.append(list(map(int, input().split())))
    total_cheese_count += sum(cheese[i])  # 전체 치즈 조각의 개수 카운트

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

time = 1
while True:
    visited = [[0] * m for _ in range(n)]
    melted_count = bfs()
    total_cheese_count -= melted_count

    if total_cheese_count == 0:  # 치즈를 다 녹였으면
        print(time)
        print(melted_count)
        break

    time += 1
```

### 1261

```python
import heapq

M, N = map(int, input().split())
maze = [list(map(int, input())) for _ in range(N)]

distance = [[float('inf')] * M for _ in range(N)]

dr = [-1, 0, 1, 0]
dc = [0, 1, 0, -1]

def dijkstra():
    q = []
    heapq.heappush(q, (0, 0, 0))
    distance[0][0] = 0

    while q:
        cost, r, c = heapq.heappop(q)

        if cost > distance[r][c]:
            continue

        for i in range(4):
            nr = r + dr[i]
            nc = c + dc[i]

            if nr < 0 or nr >= N or nc < 0 or nc >= M:
                continue

            new_cost = cost + maze[nr][nc]

            if new_cost < distance[nr][nc]:
                distance[nr][nc] = new_cost
                heapq.heappush(q, (new_cost, nr, nc))

dijkstra()

print(distance[N - 1][M - 1])
```

### 점수별 랭크

```python
def get_grade(score):
    if score >= 90:
        return "S"
    elif score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    else:
        return "F"

# 입력 받기
n = int(input().strip())

# 등급 계산 및 출력
grade = get_grade(n)
print(grade)
```
