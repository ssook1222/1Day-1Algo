```python
import sys
from collections import deque

input = sys.stdin.readline

def bfs(x, y):
    q = deque([(x, y)])
    visited[x][y] = True
    sea_list = []

    while q:
        x, y = q.popleft()
        sea = 0
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 <= nx < n and 0 <= ny < m:
                if graph[nx][ny] == 0:
                    sea += 1
                elif graph[nx][ny] > 0 and not visited[nx][ny]:
                    q.append((nx, ny))
                    visited[nx][ny] = True
        if sea > 0:
            sea_list.append((x, y, sea))

    for x, y, sea in sea_list:
        graph[x][y] = max(0, graph[x][y] - sea)

    return 1


n, m = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n)]

ice = []
for i in range(n):
    for j in range(m):
        if graph[i][j] > 0:
            ice.append((i, j))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
year = 0

while ice:
    visited = [[False] * m for _ in range(n)]
    del_list = []
    group = 0
    for i, j in ice:
        if graph[i][j] > 0 and not visited[i][j]:
            group += bfs(i, j)
        if graph[i][j] == 0:
            del_list.append((i, j))

    if group > 1:
        print(year)
        break

    ice = sorted(list(set(ice) - set(del_list)))
    year += 1

if group < 2:
    print(0)
```

### 3055

```python
import sys
from collections import deque

R, C = map(int, input().split())
forest = [list(sys.stdin.readline().rstrip()) for _ in range(R)]

visited = [[-1] * C for _ in range(R)]
dy = [-1, 1, 0, 0]
dx = [0, 0, -1, 1]

q = deque()

for y in range(R):
    for x in range(C):
        if forest[y][x] == '*':  # Water positions are added to the front
            q.appendleft((y, x))
        elif forest[y][x] == 'S':  # Hedgehog's initial position
            q.append((y, x))
            visited[y][x] = 0  # Start point with time 0

while q:
    y, x = q.popleft()
    cur = forest[y][x]  # Current position type
    for i in range(4):
        ny, nx = y + dy[i], x + dx[i]  # Next position

        if ny < 0 or ny >= R or nx < 0 or nx >= C:  # Out of bounds
            continue
        if visited[ny][nx] != -1:  # Already visited
            continue
        if forest[ny][nx] == "*":  # Water
            continue
        if forest[ny][nx] == "X":  # Rocks
            continue
        if cur == "*" and forest[ny][nx] == "D":  # Water cannot move to Beaver's den
            continue

        if cur == "S":
            if forest[ny][nx] == "D":  # Hedgehog reaches the Beaver's den
                print(visited[y][x] + 1)
                sys.exit(0)
            visited[ny][nx] = visited[y][x] + 1  # Record arrival time if not the den

        forest[ny][nx] = cur  # Update next position with hedgehog or water
        q.append((ny, nx))  # Add next position for exploration

print("KAKTUS")
```

### 길이 변형 후 사각형 넓이 구하기 2

```python
def main():
    # 입력 받기
    width, height = map(int, input().split())

    # 가로의 길이를 4배 증가시키기
    new_width = width * 4

    # 세로의 길이를 3 증가시키기
    new_height = height + 3

    # 직사각형의 넓이 계산
    area = new_width * new_height

    # 결과 출력
    print(new_width)
    print(new_height)
    print(area)

if __name__ == "__main__":
    main()
```
