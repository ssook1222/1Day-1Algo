```python
import sys
from collections import deque
from itertools import permutations

directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]

def bfs(start_r, start_c, h, w, maps):
    visited = [[0] * w for _ in range(h)]
    visited[start_r][start_c] = 1
    queue = deque([(start_r, start_c)])

    while queue:
        r, c = queue.popleft()
        for dr, dc in directions:
            nr, nc = r + dr, c + dc
            if 0 <= nr < h and 0 <= nc < w and not visited[nr][nc]:
                if maps[nr][nc] != 'x':
                    queue.append((nr, nc))
                    visited[nr][nc] = visited[r][c] + 1

    return visited

while True:
    w, h = map(int, sys.stdin.readline().split())
    if w + h:
        maps = [list(sys.stdin.readline().strip()) for _ in range(h)]
        dirty_positions = []  # 더러운 칸의 좌표를 담을 배열
        robot_r, robot_c = 0, 0  # 로봇 청소기의 위치(행, 열)

        for r in range(h):
            for c in range(w):
                if maps[r][c] == 'o':
                    robot_r, robot_c = r, c
                elif maps[r][c] == '*':
                    dirty_positions.append((r, c))

        cleaner_to_first_dirty = [0] * len(dirty_positions)  # 로봇 청소기 ~ 첫 번째로 청소할 더러운 칸까지의 거리를 담을 배열
        visited = bfs(robot_r, robot_c, h, w, maps)
        possible = True  # 로봇 청소기가 모든 더러운 칸을 방문할 수 있는지 확인

        for idx, (r, c) in enumerate(dirty_positions):
            temp = visited[r][c]
            if not temp:  # 로봇 청소기가 방문할 수 없는 칸이 나오면 False
                print(-1)
                possible = False
                break
            cleaner_to_first_dirty[idx] += temp - 1

        if possible:
            distances = [[0] * len(dirty_positions) for _ in range(len(dirty_positions))]  # 더러운 칸마다의 거리를 계산한다.

            for i in range(len(dirty_positions) - 1):
                visited = bfs(dirty_positions[i][0], dirty_positions[i][1], h, w, maps)
                for j in range(i + 1, len(dirty_positions)):
                    distances[i][j] = visited[dirty_positions[j][0]][dirty_positions[j][1]] - 1
                    distances[j][i] = distances[i][j]

            answer = int(1e9)
            for perm in permutations(range(len(distances))):  # 순열을 이용하여 최적의 순서를 찾는다.
                temp = cleaner_to_first_dirty[perm[0]]
                start = perm[0]
                for idx in range(1, len(perm)):  # 순열을 기준으로 시작점(start), 도착점(end)을 바꿔가며 거리를 더한다.
                    end = perm[idx]
                    temp += distances[start][end]
                    start = end
                answer = min(answer, temp)  # answer에 이동횟수의 최소값을 담는다.

            print(answer)
    else:
        break
```

### 진수 to 진수

```python
def convert_base(n, base):
    if n == 0:
        return "0"
    result = ""
    while n > 0:
        remainder = n % base
        result = str(remainder) + result
        n //= base
    return result

# 입력 받기
import sys
input = sys.stdin.read
data = input().split()
a = int(data[0])
b = int(data[1])
n = data[2]

# a진수를 10진수로 변환
decimal_value = int(n, a)

# 10진수를 b진수로 변환
result = convert_base(decimal_value, b)

# 결과 출력
print(result)
```
