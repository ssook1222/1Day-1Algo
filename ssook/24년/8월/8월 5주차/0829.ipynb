### 1285

```python
import sys
input = sys.stdin.readline

N = int(input())
coins = [list(map(str, input().rstrip())) for _ in range(N)]

min_flipped_count = 2147000000

# 1. 행이 뒤집힌 배열 생성
coins_reversed_rows = [["H"] * N for _ in range(N)]
for i in range(N):
    for j in range(N):
        if coins_reversed_rows[i][j] == coins[i][j]:
            coins_reversed_rows[i][j] = "T"

# 2. 비트마스크로 경우의 수 탐색
for bitmask in range(1 << N):
    temp_coins = []
    for i in range(N):
        if bitmask & (1 << i):
            temp_coins.append(coins_reversed_rows[i])
        else:
            temp_coins.append(coins[i])

    # 3. 열방향 탐색으로 뒤집는 것과 안뒤집는 것 중 최대값 저장
    total_flipped_count = 0
    for j in range(N):
        count = 0
        for i in range(N):
            if temp_coins[i][j] == "T":
                count += 1
        total_flipped_count += min(count, N - count)

    # 4. 현재 답보다 작다면 최소값이므로 저장
    min_flipped_count = min(min_flipped_count, total_flipped_count)

print(min_flipped_count)
```

### 연속되는 수 4

```python
def max_increasing_subarray_length(N, numbers):
    max_length = 1
    current_length = 1  # 현재 증가하는 연속 부분 수열의 길이 (최소 길이는 1로 시작)

    for i in range(1, N):
        if numbers[i] > numbers[i - 1]:  # 현재 숫자가 이전 숫자보다 큰 경우
            current_length += 1
        else:  # 증가하는 수열이 끊긴 경우
            max_length = max(max_length, current_length)
            current_length = 1  # 새로운 증가하는 연속 부분 수열의 길이 초기화

    # 마지막 증가하는 연속 부분 수열의 길이를 한 번 더 최대 길이와 비교
    max_length = max(max_length, current_length)

    return max_length

# 입력 받기
import sys
input = sys.stdin.read
data = input().strip().split()
N = int(data[0])
numbers = list(map(int, data[1:]))

# 최대 길이 구하기
result = max_increasing_subarray_length(N, numbers)
print(result)
```
