```python
import sys

input = sys.stdin.readline

def min_cost(row, visit):
    if row == N:
        return 0

    if dp[row][visit] != -1:
        return dp[row][visit]

    min_cost_val = float('inf')
    for i in range(N):
        if (visit & (1 << i)) != 0:  # 특정 비트가 켜져 있다면 이미 할당된 일임
            continue

        min_cost_val = min(min_cost_val, min_cost(row + 1, (visit | (1 << i))) + D[row][i])

    dp[row][visit] = min_cost_val
    return dp[row][visit]

N = int(input().strip())
D = [list(map(int, input().split())) for _ in range(N)]

# dp 배열 초기화
dp = [[-1] * (1 << N) for _ in range(N)]

# 첫 번째 사람부터 시작하여 필요한 최소 비용 계산
print(min_cost(0, 0))
```

### 연속되는 수 3

```python
def max_length_of_same_sign_subarray(N, numbers):
    max_length = 0
    current_length = 1  # 현재 연속 부분 수열의 길이 (최소 길이는 1로 시작)

    for i in range(1, N):
        if numbers[i] * numbers[i - 1] > 0:  # 현재 숫자와 이전 숫자의 부호가 동일한 경우
            current_length += 1
        else:  # 부호가 다르거나 0인 경우
            max_length = max(max_length, current_length)
            current_length = 1  # 연속 부분 수열의 길이 초기화

    # 마지막 연속 부분 수열의 길이를 한 번 더 최대 길이와 비교
    max_length = max(max_length, current_length)

    return max_length

# 입력 받기
import sys
input = sys.stdin.read
data = input().strip().split()
N = int(data[0])
numbers = list(map(int, data[1:]))

# 최대 길이 구하기
result = max_length_of_same_sign_subarray(N, numbers)
print(result)
```
