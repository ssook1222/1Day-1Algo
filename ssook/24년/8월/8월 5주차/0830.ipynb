### 17435

```python
import sys
input = sys.stdin.readline

log = 18  # log 변수는 놔두고 변수명 변경

M = int(input())  # 입력 받기
function = [0] + list(map(int, input().split()))  # 함수 f의 값들

# dp 배열 초기화
dp = [[function[i]] for i in range(M + 1)]

# dp 배열 계산
for j in range(1, log + 1):
    for i in range(1, M + 1):
        dp[i].append(dp[dp[i][j-1]][j-1])

Q = int(input())  # 쿼리 개수 입력 받기
for _ in range(Q):
    n, x = map(int, input().split())
    # 쿼리에 따라 계산
    for b in range(log, -1, -1):
        if n >= 1 << b:
            n -= 1 << b
            x = dp[x][b]
    print(x)
```

### T를 초과하는 연속 부분 수열

```python
def find_max_length_subarray(nums, t):
    n = len(nums)
    max_length = 0
    left = 0
    current_length = 0

    for right in range(n):
        if nums[right] > t:
            current_length += 1
        else:
            current_length = 0

        max_length = max(max_length, current_length)

    return max_length

# 입력 받기
n, t = map(int, input().split())
nums = list(map(int, input().split()))

# 최대 길이 구하기
result = find_max_length_subarray(nums, t)
print(result)
```
