### 1005

```python
import sys
from collections import deque

t = int(sys.stdin.readline())

for _ in range(t):
    n, k = map(int, sys.stdin.readline().rstrip().split())
    delay_time = list(map(int, sys.stdin.readline().rstrip().split()))  # 각 건물의 건설 완료 시간
    graph = [[] for _ in range(n + 1)]  # 각 건물의 선행 건물 리스트
    in_degree = [0] * (n + 1)  # 각 건물의 진입 차수
    dp = [0] * (n + 1)  # 각 건물을 건설 완료하는 데 걸리는 최소 시간
    queue = deque()

    # 간선 정보 입력 및 처리
    for _ in range(k):
        a, b = map(int, sys.stdin.readline().rstrip().split())
        graph[a].append(b)  # a번 건물을 지은 후 b번 건물을 지을 수 있음
        in_degree[b] += 1  # b번 건물의 진입 차수 증가

    w = int(sys.stdin.readline().rstrip())  # 최소 시간을 알고 싶은 건물 번호

    # 위상 정렬을 위해 진입 차수가 0인 건물들을 큐에 삽입
    for i in range(1, n + 1):
        if in_degree[i] == 0:
            queue.append(i)
            dp[i] = delay_time[i - 1]  # 진입 차수가 0인 건물은 자신의 건설 완료 시간으로 초기화

    # 위상 정렬 수행
    while queue:
        current = queue.popleft()

        for next_building in graph[current]:
            in_degree[next_building] -= 1  # current를 지었으므로 next_building의 진입 차수 감소
            dp[next_building] = max(dp[next_building], dp[current] + delay_time[next_building - 1])

            if in_degree[next_building] == 0:
                queue.append(next_building)

    # 결과 출력
    print(dp[w])
```

### 1202

```python
import sys
import heapq

N, K = map(int, sys.stdin.readline().split())
jewels = []
for _ in range(N):
    heapq.heappush(jewels, list(map(int, sys.stdin.readline().split())))

bags = []
for _ in range(K):
    bags.append(int(sys.stdin.readline()))
bags.sort()

answer = 0
tmp_jewels = []

for bag in bags:
    while jewels and bag >= jewels[0][0]:
        heapq.heappush(tmp_jewels, -heapq.heappop(jewels)[1])

    if tmp_jewels:
        answer -= heapq.heappop(tmp_jewels)
    elif not jewels:
        break

print(answer)
```
