### 11689

```python
import sys

def euler_totient(n):
    result = n
    p = 2
    while p * p <= n:
        if n % p == 0:
            while n % p == 0:
                n //= p
            result -= result // p
        p += 1
    if n > 1:
        result -= result // n
    return result

N = int(sys.stdin.readline().strip())
print(euler_totient(N))
```

### 13334

```python
import sys
import heapq

input = sys.stdin.read().strip().split("\n")
n = int(input[0])
locations = []
for i in range(1, n+1):
    h, o = map(int, input[i].split())
    locations.append((min(h, o), max(h, o)))
d = int(input[n+1])

locations.sort(key=lambda x: (x[1], x[0]))

heap = []
max_people = 0

for start, end in locations:
    heapq.heappush(heap, start)
    line_start = end - d

    while heap and heap[0] < line_start:
        heapq.heappop(heap)

    max_people = max(max_people, len(heap))

print(max_people)
```
