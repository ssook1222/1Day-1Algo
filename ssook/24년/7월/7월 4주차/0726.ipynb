### 1414

```python
import sys
input = sys.stdin.readline

def find(parent, x):
    if parent[x] != x:
        parent[x] = find(parent, parent[x])
    return parent[x]

def union(parent, a, b):
    root_a = find(parent, a)
    root_b = find(parent, b)
    if root_a > root_b:
        parent[root_a] = root_b
    else:
        parent[root_b] = root_a

n = int(input())
lan_lengths = [list(input().rstrip()) for _ in range(n)]
parent = [i for i in range(n+1)]
edges = []

total_length = 0
for i in range(n):
    for j in range(n):
        if lan_lengths[i][j] == '0':
            edges.append((0, i+1, j+1))
        else:
            if 'a' <= lan_lengths[i][j] <= 'z':
                length = ord(lan_lengths[i][j]) - ord('a') + 1
                edges.append((length, i+1, j+1))
                total_length += length
            elif 'A' <= lan_lengths[i][j] <= 'Z':
                length = ord(lan_lengths[i][j]) - ord('A') + 27
                edges.append((length, i+1, j+1))
                total_length += length

edges.sort()

for length, room_a, room_b in edges:
    if length == 0:
        continue
    if find(parent, room_a) != find(parent, room_b):
        union(parent, room_a, room_b)
        total_length -= length

all_connected = True
for i in range(1, n + 1):
    if find(parent, i) != find(parent, 1):
        all_connected = False

if all_connected:
    print(total_length)
else:
    print(-1)
```

### 1460

```python
import sys
input = sys.stdin.readline

n, m = map(int, input().split())

farm = [[0] * n for _ in range(n)]

for _ in range(m):
    x, y, l, f = map(int, input().split())
    for i in range(y, y + l):
        for j in range(x, x + l):
            farm[i][j] = f

def largest_square(fruit1, fruit2):
    dp = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            if farm[i][j] == fruit1 or farm[i][j] == fruit2:
                dp[i][j] = 1

    for i in range(1, n):
        for j in range(1, n):
            if not dp[i][j]:
                continue
            dp[i][j] = min(dp[i][j - 1], dp[i - 1][j], dp[i - 1][j - 1]) + 1

    return max(map(max, dp))

max_area = 0
for fruit1 in range(1, 8):
    for fruit2 in range(fruit1 + 1, 8):
        max_area = max(max_area, largest_square(fruit1, fruit2))

print(max_area ** 2)
```
