## BoJ

### 3595

```python
n = int(input())

min_surface = float('inf')
best = (1, 1, n)

# a는 1부터 n의 세제곱근까지
for a in range(1, int(n ** (1/3)) + 2):
    if n % a == 0:
        remaining = n // a
        # b는 a부터 remaining의 제곱근까지
        for b in range(a, int(remaining ** 0.5) + 2):
            if remaining % b == 0:
                c = remaining // b
                if b <= c:  # a ≤ b ≤ c 보장
                    surface = 2 * (a*b + b*c + c*a)
                    if surface < min_surface:
                        min_surface = surface
                        best = (a, b, c)

print(best[0], best[1], best[2])
```
