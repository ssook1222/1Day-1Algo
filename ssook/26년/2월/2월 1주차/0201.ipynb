```python
# 6줄의 입력을 받음
# 줄 끝의 공백은 카운트하지 않음
lines = []
for _ in range(6):
    line = input().rstrip()
    lines.append(len(line))

# 첫 3줄은 위도 (Latitude)
latitude = lines[0:3]

# 마지막 3줄은 경도 (Longitude)
longitude = lines[3:6]

# 출력
print(f"Latitude {latitude[0]}:{latitude[1]}:{latitude[2]}")
print(f"Longitude {longitude[0]}:{longitude[1]}:{longitude[2]}")
```
