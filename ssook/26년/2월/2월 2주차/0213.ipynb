## BoJ

### 27797

```python
t = int(input())

for case_num in range(1, t + 1):
    n = int(input())
    matrix = []

    # 행렬 입력 받기
    for i in range(n):
        row = list(map(int, input().split()))
        matrix.append(row)

    # 1. Trace 계산 (대각선 합)
    trace = 0
    for i in range(n):
        trace += matrix[i][i]

    # 2. 중복된 값을 가진 행의 개수
    repeated_rows = 0
    for i in range(n):
        row = matrix[i]
        if len(row) != len(set(row)):  # 중복이 있으면
            repeated_rows += 1

    # 3. 중복된 값을 가진 열의 개수
    repeated_cols = 0
    for j in range(n):
        col = [matrix[i][j] for i in range(n)]
        if len(col) != len(set(col)):  # 중복이 있으면
            repeated_cols += 1

    print(f"Case #{case_num}: {trace} {repeated_rows} {repeated_cols}")
```
