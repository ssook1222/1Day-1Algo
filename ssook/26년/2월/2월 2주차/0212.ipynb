```python
def count_divisors(num):
    """주어진 수의 약수 개수를 계산"""
    count = 0
    i = 1
    while i * i <= num:
        if num % i == 0:
            count += 1
            if i * i != num:
                count += 1
        i += 1
    return count

n, k = map(int, input().split())

# n×n 테이블 생성
table = []
num = 1

for i in range(n):
    row = []
    for j in range(n):
        # num의 약수 개수가 k 이하면 *, 아니면 .
        if count_divisors(num) <= k:
            row.append('*')
        else:
            row.append('.')
        num += 1
    table.append(''.join(row))

# 출력
for row in table:
    print(row)
```
