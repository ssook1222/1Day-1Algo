## BoJ

### 6246

```python
n, q = map(int, input().split())

# 슬롯 상태 (False: 비어있음, True: 풍선 있음)
slots = [False] * (n + 1)  # 1번부터 N번까지 사용 (0번 인덱스는 사용 안 함)

for _ in range(q):
    l, i = map(int, input().split())

    # L번 슬롯부터 I개씩 띄어서 풍선 설치
    position = l
    while position <= n:
        # 이미 풍선이 없는 슬롯에만 설치
        if not slots[position]:
            slots[position] = True
        position += i

# 비어있는 슬롯 개수 세기 (1번부터 N번까지)
empty_count = sum(1 for i in range(1, n + 1) if not slots[i])

print(empty_count)
```
