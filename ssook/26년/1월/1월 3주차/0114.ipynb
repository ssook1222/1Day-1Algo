```python
# 입력 받기
T, N = map(int, input().split())

# 요일을 시간으로 변환하는 딕셔너리 (월요일 0시를 기준으로)
day_to_hours = {
    'Mon': 0,
    'Tue': 24,
    'Wed': 48,
    'Thu': 72,
    'Fri': 96
}

# 평일 동안 잔 총 시간 계산
total_sleep = 0

for _ in range(N):
    line = input().split()
    d1, h1, d2, h2 = line[0], int(line[1]), line[2], int(line[3])

    # 잠든 시각을 절대 시간으로 변환 (월요일 0시 기준)
    sleep_start = day_to_hours[d1] + h1
    sleep_end = day_to_hours[d2] + h2

    # 수면 시간 계산
    sleep_duration = sleep_end - sleep_start
    total_sleep += sleep_duration

# 주말에 자야 할 시간 계산
need_to_sleep = T - total_sleep

# 결과 출력
if need_to_sleep <= 0:
    print(0)
elif need_to_sleep > 48:
    print(-1)
else:
    print(need_to_sleep)
```
