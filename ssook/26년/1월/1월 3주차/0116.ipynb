## BoJ

### 3553

```python
# 입력 받기
n, d = map(int, input().split())

# n자리 수의 범위
if n == 1:
    min_num = 1
else:
    min_num = 10 ** (n - 1)

max_num = 10 ** n - 1

# min_num 이상에서 d로 나누어떨어지는 첫 번째 수 찾기
if min_num % d == 0:
    answer = min_num
else:
    # min_num보다 큰 d의 배수 찾기
    answer = ((min_num // d) + 1) * d

# 답이 n자리 범위 내에 있는지 확인
if answer <= max_num:
    print(answer)
else:
    print("No solution")
```
