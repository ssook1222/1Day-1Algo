```python
# 입력 받기
k, n = map(int, input().split())

# 최소 시간: 한 캔을 집중적으로 먹어서 빨리 비우기
# n개를 먹고 나서 그 빈 캔을 다시 집는 경우
min_time = n + 1

# 최대 시간: 모든 칩을 다 먹고 나서 빈 캔을 집는 경우
# k * n개를 모두 먹고 나서 빈 캔을 집는 경우
max_time = k * n + 1

# 결과 출력
print(min_time, max_time)
```
