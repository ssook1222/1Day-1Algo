```python
# 입력 받기
A, B = input().split()

# A에서 B에도 포함된 첫 번째 글자 찾기
cross_char = None
pos_a = -1
pos_b = -1

for i, char in enumerate(A):
    if char in B:
        cross_char = char
        pos_a = i
        pos_b = B.index(char)  # B에서 처음 나오는 위치
        break

# 크로스워드 출력
# B의 길이만큼 줄을 출력
for i in range(len(B)):
    if i == pos_b:
        # 교차하는 줄 - A를 출력
        print(A)
    else:
        # B의 i번째 글자를 pos_a 위치에 출력
        line = '.' * pos_a + B[i] + '.' * (len(A) - pos_a - 1)
        print(line)
```
