```python
# 입력 받기
N = int(input())
data_bytes = []
for _ in range(N):
    byte_data = list(map(int, input().split()))
    data_bytes.append(byte_data)

parity_byte = list(map(int, input().split()))

# 1단계: 패리티 타입 결정 (손상되지 않은 바이트들로부터)
# 각 바이트의 1의 개수를 세어봄
parity_counts = []
for i in range(N):
    count = sum(data_bytes[i])
    parity_counts.append(count % 2)  # 0: even, 1: odd

# 가장 많이 나타나는 패리티가 올바른 패리티
# (N개 중 1개만 손상되었으므로 나머지는 올바름)
odd_count = sum(parity_counts)
even_count = N - odd_count

if odd_count > even_count:
    parity_type = "Odd"
    correct_parity = 1
else:
    parity_type = "Even"
    correct_parity = 0

# 2단계: 손상된 바이트 찾기 (행 검사)
broken_byte = -1
for i in range(N):
    if parity_counts[i] != correct_parity:
        broken_byte = i + 1  # 1-indexed
        break

# 3단계: 손상된 비트 찾기 (열 검사)
# 각 열(비트 위치)의 패리티를 검사
broken_bit = -1
for bit_pos in range(8):
    # 해당 열의 모든 데이터 바이트 비트를 합산
    column_sum = sum(data_bytes[i][bit_pos] for i in range(N))

    # 패리티 바이트의 해당 비트
    parity_bit = parity_byte[bit_pos]

    # 패리티 검사: (데이터 합 + 패리티 비트) % 2 == correct_parity
    total_parity = (column_sum + parity_bit) % 2

    if total_parity != correct_parity:
        broken_bit = bit_pos + 1  # 1-indexed
        break

# 결과 출력
print(parity_type)
print(f"Byte {broken_byte} is broken")
print(f"Bit {broken_bit} is broken")
```
