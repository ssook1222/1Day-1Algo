```python
# 첫 번째 온도 입력
previous_temp = float(input())

# 계속해서 온도 입력받기
while True:
    temp = float(input())

    # 999가 입력되면 종료
    if temp == 999:
        break

    # 온도 변화 계산 (현재 - 이전)
    change = temp - previous_temp

    # 소수점 둘째자리까지 출력
    print(f"{change:.2f}")

    # 현재 온도를 이전 온도로 업데이트
    previous_temp = temp
```
