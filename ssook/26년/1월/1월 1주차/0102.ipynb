## BoJ

### 6526

```python
while True:
    n = int(input())
    if n == 0:
        break

    perm = list(map(int, input().split()))

    # 순열이 ambiguous한지 확인
    # p[i]번째 위치에 i가 있어야 함 (1-indexed)
    # 즉, perm[perm[i-1] - 1] == i 가 모든 i에 대해 성립해야 함

    is_ambiguous = True
    for i in range(n):
        # i+1이 perm[i]번째 위치에 있는지 확인
        # perm[i]번째 위치는 perm[perm[i]-1]
        if perm[perm[i] - 1] != i + 1:
            is_ambiguous = False
            break

    if is_ambiguous:
        print("ambiguous")
    else:
        print("not ambiguous")
```
