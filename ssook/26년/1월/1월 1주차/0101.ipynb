## BoJ

### 3588

```python
# 입력 받기
K = int(input())

for data_set in range(1, K + 1):
    n = int(input())  # 수도관 파열 개수
    s, f = map(int, input().split())  # 계산할 시간 구간

    # 수도관 파열 정보 저장
    breaks = []
    for _ in range(n):
        si, fi, ri = map(int, input().split())
        breaks.append((si, fi, ri))

    # 각 시간마다 낭비되는 물의 양 계산
    total_waste = 0
    for t in range(s, f + 1):  # s부터 f까지 포함
        for si, fi, ri in breaks:
            # 시간 t에 이 파열이 진행 중이면
            if si <= t <= fi:
                total_waste += ri

    # 출력
    print(f"Data Set {data_set}:")
    print(total_waste)
    print()  # 빈 줄
```
