```python
import math

# 첫 번째 플레이어의 주사위
dice1 = list(map(int, input().split()))

# 두 번째 플레이어의 주사위
dice2 = list(map(int, input().split()))

# 첫 번째 플레이어가 이기는 경우의 수 계산
win_count = 0

for face1 in dice1:
    for face2 in dice2:
        if face1 > face2:
            win_count += 1

# 전체 경우의 수
total_count = 36

# 최대공약수로 약분
gcd = math.gcd(win_count, total_count)
numerator = win_count // gcd
denominator = total_count // gcd

print(f"{numerator}/{denominator}")
```
