```python
def max_consecutive_length(capacities):
    """주어진 리스트에서 연속된 같은 값의 최대 길이를 반환"""
    if not capacities:
        return 0

    max_len = 1
    current_len = 1

    for i in range(1, len(capacities)):
        if capacities[i] == capacities[i-1]:
            current_len += 1
            max_len = max(max_len, current_len)
        else:
            current_len = 1

    return max_len

# 입력 받기
n = int(input())
capacities = []
for _ in range(n):
    capacities.append(int(input()))

# 모든 서로 다른 용량 찾기
unique_capacities = set(capacities)

# 각 용량을 제거했을 때의 최대 연속 길이 계산
max_result = 0

for remove_capacity in unique_capacities:
    # 해당 용량을 제거한 리스트 생성
    remaining = [c for c in capacities if c != remove_capacity]

    # 남은 리스트에서 최대 연속 길이 계산
    if remaining:  # 리스트가 비어있지 않으면
        length = max_consecutive_length(remaining)
        max_result = max(max_result, length)

print(max_result)
```
