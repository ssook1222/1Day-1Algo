```python
import sys

def last_nonzero_digit(n):
    if n < 2:
        return 1

    # 2의 개수와 5의 개수를 센다
    count_2 = 0
    count_5 = 0

    # 끝자리만 유지하면서 곱셈
    result = 1

    for i in range(2, n + 1):
        num = i

        # 2를 모두 제거
        while num % 2 == 0:
            count_2 += 1
            num //= 2

        # 5를 모두 제거
        while num % 5 == 0:
            count_5 += 1
            num //= 5

        # 2와 5가 제거된 수를 곱함 (끝자리만 유지)
        result = (result * num) % 10

    # 2와 5는 쌍으로 10을 만들어 0이 됨
    # 남은 2들을 곱해야 함
    remaining_2 = count_2 - count_5

    # 2를 remaining_2번 곱함 (끝자리만)
    for _ in range(remaining_2):
        result = (result * 2) % 10

    return result

# 입력 처리
for line in sys.stdin:
    n = int(line.strip())
    last_digit = last_nonzero_digit(n)
    print(f"{n:>5} -> {last_digit}")
```
