```python
# 입력 받기
W = int(input())  # 도로 길이 (마일)
N = int(input())  # 물총 사이 거리 (피트)

# 마일을 피트로 변환
total_feet = W * 5280

# 물총에 맞는 횟수 계산
# N 피트마다 물총이 있으므로, total_feet를 N으로 나눈 몫이 답
water_gun_count = total_feet // N

# 결과 출력
print(water_gun_count)
```
