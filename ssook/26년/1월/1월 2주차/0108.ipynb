```python
# 입력 받기
r, c = map(int, input().split())
grid = []
for _ in range(r):
    row = list(map(int, input().split()))
    grid.append(row)

# 결과를 저장할 배열
result = []

# 각 셀에 대해 확인
for i in range(r):
    row_result = []
    for j in range(c):
        # 가장자리는 배수구 불필요
        if i == 0 or i == r - 1 or j == 0 or j == c - 1:
            row_result.append(0)
        else:
            # 상하좌우 4개 방향 확인
            current = grid[i][j]
            up = grid[i - 1][j]
            down = grid[i + 1][j]
            left = grid[i][j - 1]
            right = grid[i][j + 1]

            # 4개 방향이 모두 현재보다 높으면 배수구 필요
            if up > current and down > current and left > current and right > current:
                row_result.append(1)
            else:
                row_result.append(0)

    result.append(row_result)

# 출력
for row in result:
    print(' '.join(map(str, row)))
```
