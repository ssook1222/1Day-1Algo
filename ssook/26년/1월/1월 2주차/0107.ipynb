## BoJ

### 1233

```python
# 입력 받기
S1, S2, S3 = map(int, input().split())

# 합의 빈도를 저장할 딕셔너리
frequency = {}

# 세 주사위의 모든 조합 확인
for i in range(1, S1 + 1):
    for j in range(1, S2 + 1):
        for k in range(1, S3 + 1):
            dice_sum = i + j + k
            if dice_sum in frequency:
                frequency[dice_sum] += 1
            else:
                frequency[dice_sum] = 1

# 가장 높은 빈도 찾기
max_frequency = max(frequency.values())

# 가장 높은 빈도를 가진 합 중 가장 작은 값 찾기
result = min([dice_sum for dice_sum, freq in frequency.items() if freq == max_frequency])

print(result)
```
