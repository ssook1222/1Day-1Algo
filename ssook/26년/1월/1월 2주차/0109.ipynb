## BoJ

### 5939

```python
# 입력 받기
N = int(input())
times = []

for _ in range(N):
    h, m, s = map(int, input().split())
    times.append((h, m, s))

# 시간 정렬 (시, 분, 초 순서로 자동 정렬됨)
times.sort()

# 출력
for h, m, s in times:
    print(h, m, s)
```
