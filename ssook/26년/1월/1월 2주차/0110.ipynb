```python
# 입력 받기
n = int(input())

# 럭키 번호 개수 계산
# XXX - XX - XX = 0
# 즉, n = a + b를 만족하는 (a, b) 쌍의 개수
# 여기서 0 ≤ a, b ≤ 99

count = 0

# a를 0부터 99까지 시도
for a in range(100):
    b = n - a
    # b도 0부터 99 사이에 있어야 함
    if 0 <= b <= 99:
        count += 1

print(count)
```
