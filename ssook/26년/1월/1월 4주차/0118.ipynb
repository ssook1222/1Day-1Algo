```python
T = int(input())

for _ in range(T):
    N = int(input())

    # 각 플레이어가 지목한 사람 (1-indexed)
    next_player = [0]  # 0번 인덱스는 사용 안 함
    for i in range(N):
        next_player.append(int(input()))

    # 희현이(1번)부터 시작해서 경로 추적
    current = 1
    visited = {}  # {플레이어 번호: 몇 번째 방문}
    step = 0

    # 주경이(N번)를 만날 때까지 또는 사이클을 발견할 때까지
    while True:
        # 현재 플레이어가 이미 방문했다면 사이클 발견
        if current in visited:
            # 주경이를 한 번도 방문하지 않았다면 불가능
            if N not in visited:
                print(0)
            else:
                # 주경이를 이미 방문했다면 그 때의 K 출력
                print(visited[N])
            break

        # 현재 위치 기록
        visited[current] = step

        # 주경이(N번)를 처음 만났다면
        if current == N and step > 0:
            print(step)
            break

        # 다음 플레이어로 이동
        current = next_player[current]
        step += 1
```
