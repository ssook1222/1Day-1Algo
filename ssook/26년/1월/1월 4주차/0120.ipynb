## BoJ

### 11575

```python
T = int(input())

for _ in range(T):
    a, b = map(int, input().split())
    s = input().strip()

    result = []
    for char in s:
        # 알파벳을 숫자로 변환 (A=0, B=1, ..., Z=25)
        x = ord(char) - ord('A')

        # 아핀 암호 공식 적용: E(x) = (ax + b) mod 26
        encrypted = (a * x + b) % 26

        # 숫자를 다시 알파벳으로 변환
        encrypted_char = chr(encrypted + ord('A'))
        result.append(encrypted_char)

    print(''.join(result))
```
