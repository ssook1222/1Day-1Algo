```python
import math

# 입력 받기
n, s = map(int, input().split())
times = list(map(int, input().split()))

# 가장 느린 솔루션 찾기
slowest = max(times)

# s배 계산 (밀리초)
required_ms = slowest * s

# 초 단위로 변환 (올림)
required_seconds = math.ceil(required_ms / 1000)

print(required_seconds)
```
