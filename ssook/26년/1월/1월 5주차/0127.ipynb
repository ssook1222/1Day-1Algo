## BoJ

### 14209

```python
n = int(input())

# 카드 점수 딕셔너리
points = {
    'A': 4,
    'K': 3,
    'Q': 2,
    'J': 1,
    'X': 0
}

total = 0

for _ in range(n):
    cards = input()
    for card in cards:
        total += points[card]

print(total)
```
