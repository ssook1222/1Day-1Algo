```python
import math

def parse_height(height_str):
    """키를 파싱하여 총 인치로 변환"""
    parts = height_str.split("'")
    feet = int(parts[0])
    inches = int(parts[1].rstrip('"'))
    return feet * 12 + inches

def format_height(total_inches):
    """총 인치를 피트'인치" 형식으로 변환"""
    feet = total_inches // 12
    inches = total_inches % 12
    return f"{feet}'{inches}\""

T = int(input())

for case_num in range(1, T + 1):
    line = input().split()
    gender = line[0]
    mother_height = parse_height(line[1])
    father_height = parse_height(line[2])

    # 부모 키의 합
    parent_sum = mother_height + father_height

    # 성별에 따라 5인치 더하거나 빼기
    if gender == 'B':  # 남자아이
        adjusted_sum = parent_sum + 5
    else:  # 여자아이 (G)
        adjusted_sum = parent_sum - 5

    # 2로 나누기
    mid_value = adjusted_sum / 2

    # 4인치 범위
    min_height = mid_value - 4
    max_height = mid_value + 4

    # 정수가 아니면 안쪽으로 조이기 (shrink)
    min_height_int = math.ceil(min_height)
    max_height_int = math.floor(max_height)

    # 출력
    print(f"Case #{case_num}: {format_height(min_height_int)} to {format_height(max_height_int)}")
```
