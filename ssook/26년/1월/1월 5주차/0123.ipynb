## BoJ

### 11038

```python
while True:
    line = input().split()
    m, nmin, nmax = map(int, line)

    # 종료 조건
    if m == 0 and nmin == 0 and nmax == 0:
        break

    # 점수 읽기
    scores = []
    for _ in range(m):
        scores.append(int(input()))

    # 최대 gap과 그때의 n 찾기
    max_gap = -1
    best_n = nmin

    for n in range(nmin, nmax + 1):
        # n명을 합격시킬 때의 gap 계산
        # 합격자 최저 점수: scores[n-1] (0-indexed이므로)
        # 불합격자 최고 점수: scores[n]
        gap = scores[n - 1] - scores[n]

        # gap이 더 크거나, 같으면서 n이 더 크면 업데이트
        if gap > max_gap or (gap == max_gap and n > best_n):
            max_gap = gap
            best_n = n

    print(best_n)
```
