## BoJ

### 12540

```python
n = int(input())

for case_num in range(1, n + 1):
    m = int(input())
    prices = list(map(int, input().split()))

    max_profit = -1
    best_buy = -1
    best_sell = -1
    best_buy_price = float('inf')

    # 모든 구매-판매 조합 확인
    for buy_month in range(11):  # 0부터 10까지 (1월~11월)
        buy_price = prices[buy_month]
        quantity = m // buy_price  # 구매 가능한 수량

        if quantity == 0:
            continue

        for sell_month in range(buy_month + 1, 12):  # 구매 다음 달부터 12월까지
            sell_price = prices[sell_month]

            # 이익 계산
            profit = (sell_price - buy_price) * quantity

            # 최대 이익 업데이트
            # 같은 이익이면 구매가가 낮은 것 선택
            if profit > max_profit or (profit == max_profit and buy_price < best_buy_price):
                max_profit = profit
                best_buy = buy_month + 1  # 1-indexed
                best_sell = sell_month + 1  # 1-indexed
                best_buy_price = buy_price

    # 결과 출력
    if max_profit <= 0:
        print(f"Case #{case_num}: IMPOSSIBLE")
    else:
        print(f"Case #{case_num}: {best_buy} {best_sell} {max_profit}")
```
