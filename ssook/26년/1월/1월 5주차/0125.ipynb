```python
def count_non_moo(x):
    multiples_of_3 = x // 3
    multiples_of_5 = x // 5
    multiples_of_15 = x // 15

    return x - multiples_of_3 - multiples_of_5 + multiples_of_15


def find_nth_number(n):
    left, right = 1, 2 * n  # 충분히 큰 범위

    while left < right:
        mid = (left + right) // 2

        # mid까지 카운트했을 때 "Moo"가 아닌 숫자의 개수
        count = count_non_moo(mid)

        if count < n:
            # 아직 부족함, 더 큰 숫자 필요
            left = mid + 1
        else:
            # mid가 너무 크거나 정확함
            right = mid

    return left


# 입력
n = int(input())

# N번째 숫자 찾기
result = find_nth_number(n)

print(result)
```
