```python
def calculate_distance(X, Y, n):
    max_distance = 0

    # For each element in X
    for i in range(n):
        # Find the largest j >= i where Y[j] >= X[i]
        largest_j = -1
        for j in range(i, n):
            if Y[j] >= X[i]:
                largest_j = j

        # If we found a valid j, calculate distance
        if largest_j >= i:
            distance = largest_j - i
            max_distance = max(max_distance, distance)

    return max_distance


def calculate_distance_optimized(X, Y, n):
    max_distance = 0
    j = n - 1  # Start from the end of Y

    # For each element in X
    for i in range(n):
        # Move j backwards while Y[j] < X[i]
        while j >= i and Y[j] < X[i]:
            j -= 1

        # If we found a valid j
        if j >= i:
            distance = j - i
            max_distance = max(max_distance, distance)

    return max_distance


# Read number of test cases
T = int(input())

for test_case in range(T):
    # Read sequence length
    n = int(input())

    # Read X sequence
    X = list(map(int, input().split()))

    # Read Y sequence
    Y = list(map(int, input().split()))

    # Calculate distance
    max_dist = calculate_distance(X, Y, n)

    # Output result
    print(f"The maximum distance is {max_dist}")

    # Print blank line after each test case except the last one
    if test_case < T - 1:
        print()
```
