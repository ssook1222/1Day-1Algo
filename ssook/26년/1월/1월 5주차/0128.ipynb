## BoJ

### 31496

```python
n, s = input().split()
n = int(n)

total_deleted = 0

for _ in range(n):
    line = input().split()
    item_name = line[0]
    quantity = int(line[1])

    # 아이템 이름을 '_'로 분리하여 단어 리스트 생성
    words = item_name.split('_')

    # 단어 리스트에 s가 포함되어 있는지 확인
    if s in words:
        total_deleted += quantity

print(total_deleted)
```
