```python
while True:
    line = input().split()
    n, m, c = map(int, line)

    # 종료 조건
    if n == 0 and m == 0 and c == 0:
        break

    possible_positions = (n - 7) * (m - 7)

    bottom_right_parity = (n - 1 + m - 1) % 2

    target_parity = (n + m - 16) % 2
    if c == 0:
        target_parity = 1 - target_parity

    rows = n - 7
    cols = m - 7

    even_rows = (rows + 1) // 2
    odd_rows = rows // 2
    even_cols = (cols + 1) // 2
    odd_cols = cols // 2

    if target_parity == 0:
        count = even_rows * even_cols + odd_rows * odd_cols
    else:
        count = even_rows * odd_cols + odd_rows * even_cols

    print(count)
```
