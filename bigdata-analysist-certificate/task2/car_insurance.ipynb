## 자동차 보험가입 예측데이터

### 라이브러리 & 패키지 다 불러오기

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.ensemble import RandomForestClassifier # 계속 뻑나서 그냥 의사결정나무로 돌림
from sklearn import tree
```

```python
from sklearn.metrics import accuracy_score
```

### 데이터 다 불러오기

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/y_test.csv')
```

### 데이터 전처리

#### 데이터 결측치 확인

```python
x_train[(x_train.isna()==True)].count()
```

없음. 바로 라벨 인코딩 진행

```python
x_train.info()
```

```python
le = LabelEncoder()
```

```python
x_train['Gender'] = le.fit_transform(x_train['Gender'])
x_train['Vehicle_Age'] = le.fit_transform(x_train['Vehicle_Age'])
x_train['Vehicle_Damage'] = le.fit_transform(x_train['Vehicle_Damage'])

x_test['Gender'] = le.fit_transform(x_test['Gender'])
x_test['Vehicle_Age'] = le.fit_transform(x_test['Vehicle_Age'])
x_test['Vehicle_Damage'] = le.fit_transform(x_test['Vehicle_Damage'])
```

```python
display(x_train)
display(x_test)
```

수치 차이가 너무 큼. min-max 정규화 진행

### Min-Max 정규화

```python
min_max = MinMaxScaler()
```

```python
x_train_data = min_max.fit_transform(x_train)
x_train = pd.DataFrame(data=x_train_data, columns=x_train.columns)

x_test_data = min_max.fit_transform(x_test)
x_test = pd.DataFrame(data=x_test_data, columns=x_test.columns)
```

```python
display(x_train)
display(x_test)
```

### 데이터 모델링 진행

```python
x_train.columns
x_train = x_train[['ID', 'Gender', 'Age', 'Driving_License', 'Region_Code',
       'Previously_Insured', 'Vehicle_Age', 'Vehicle_Damage', 'Annual_Premium',
       'Policy_Sales_Channel', 'Vintage']]
```

```python
x_test
x_test = x_test[['ID', 'Gender', 'Age', 'Driving_License', 'Region_Code',
       'Previously_Insured', 'Vehicle_Age', 'Vehicle_Damage', 'Annual_Premium',
       'Policy_Sales_Channel', 'Vintage']]
```

```python
x_train = x_train[:100]
y_train = y_train[:100]

x_test = x_test[:100]
y_test = y_test[:100]
```

```python
rfc = RandomForestClassifier(n_estimators=1, max_depth=3, n_jobs=-1, max_samples=1, random_state=0)
```

```python
model = rfc.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```

```python
y_test
```

```python
print(accuracy_score(predict[:,1], y_test['Response']))
```

### 제출하기

```python
submit = pd.DataFrame({'id':y_test['ID'], 'Response':predict[:,1]})
submit.to_csv('result.csv', index = False)
```
