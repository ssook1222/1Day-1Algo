### 1-2

BMI는 몸무게(kg) / (키(M) * 키(M)) 로 정의 된다. 초고도 비만은 BMI 25이상 , 고도 비만은 BMI 25미만 - 23이상 , 정상은 23미만 - 18.5이상 저체중은 18.5미만으로 정의 된다. 주어진 데이터에서 초고도비만 인원 + 저체중 인원 의 숫자는?

```python
print(df.columns)
```

```python
df['BMI'] = df['weight(kg)'] / (((df['height(cm)']*0.01) * (df['height(cm)']*0.01)))
```

```python
biman = 0
malla = 0

def chk(x):
  global biman
  global malla

  if x >= 25:
    biman += 1

  elif x < 18.5:
    malla += 1
```

```python
df.BMI.apply(chk)

print(biman + malla)
```

## 작업 2유형

예측 변수 price, test.csv에 대해 ID별로 price 값을 예측하여 제출,       
제출 데이터 컬럼은 ID와 price 두개만 존재해야함.       
평가지표는 rmse     

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_test_.csv')

display(train.head(2))
test.head(2)
```

### 패키지 다 때려 박기

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.metrics import mean_absolute_error
```

### 데이터 생성

```python
print(train.columns)
```

```python
x_train = train[['model', 'year', 'transmission', 'mileage', 'fuelType',
       'tax', 'mpg', 'engineSize']]
x_test = test[['model', 'year', 'transmission', 'mileage', 'fuelType',
       'tax', 'mpg', 'engineSize']]
```

```python
y_train = train[['ID', 'price']]
```

### 데이터 전처리 진행

```python
print(x_train.isna().sum())
print(x_test.isna().sum())
```

```python
print(x_train.info())
print(x_test.info())
```

```python
print(x_train.columns)
```

```python
columns = ['model', 'transmission', 'fuelType']
le = LabelEncoder()

for column in columns:
  x_test[column] = le.fit_transform(x_test[column])
  x_train[column] = le.fit_transform(x_train[column])
```

### 데이터 모델링

```python
dtr = DecisionTreeRegressor()
model = dtr.fit(x_train, y_train)
predict1 = model.predict(x_test)
```

### 제출하기

```python
res = pd.DataFrame(columns = y_train.columns, data = predict1)
display(res.head(5))
res.to_csv('result.csv')
```

## 작업 3유형

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p3_1.csv')
df.head()
```

이 학생들의 키의 95% 신뢰구간을 구하고자 한다.     

----

### 3-1-a

55명 학생들의 키에 대한 표본 평균을 구하여라(반올림하여 소숫점 3째자리까지)

```python
import numpy as np
```

```python
height = df['height'].mean()
```

```python
print(round(height, 3))
```

---

### 3-1-b

t분포 양쪽 꼬리에서의 t 값을 구하여라 (반올림하여 소수4째자리까지)

! b ~ c 다시 한 번 풀어보기

```python
import numpy as np
from scipy.stats import t # scipy 외우기...
```

```python
std = np.std(df.height, ddof=1) # 표준편차 구하고
n = len(df.height) # 자유도 구하기

# 신뢰수준, 자유도
confidence_level = 0.95
ddof = n - 1

# t 분포의 양쪽 꼬리에서의 t값
t_value = round(t.ppf((1 + confidence_level) / 2, ddof),4)
print(t_value)
```

---

### 3-1-c

95% 신뢰구간을 구하여라(print(lower,upper) 방식으로 출력, 각각의 값은 소숫점 이하 3째자리까지)

```python
mean = df['height'].mean()
```

```python
lower = round(mean - t_value * std / np.sqrt(n),3)
upper = round(mean + t_value * std / np.sqrt(n),3)
print(lower,upper)
```

### 3-1-a

3 그룹의 데이터에 대해 크루스칼-왈리스 검정을 사용하여 검정 통계량을 반올림하여 소숫점 이하 3자리까지 구하여라

```python
import scipy.stats
```

```python
dir(scipy.stats)
```

```python
from scipy.stats import kruskal
```

아래처럼 그룹별로 파싱

```python
a = df[df.ID =='A'].value.values
b = df[df.ID =='B'].value.values
c = df[df.ID =='C'].value.values
```

```python
s,p = kruskal(a,b,c)

round_s = round(s,3)
print(round_s)
```

### 3-1-b

3 그룹의 데이터에 대해 크루스칼-왈리스 검정을 사용하여 p-value를 반올림하여 소숫점 이하 3자리까지 구하여라. 귀무가설과 대립가설중 0.05 유의수준에서 유의한 가설을 출력하라

```python
print(round(p, 3))
```

```python
if p > 0.05:
  print("귀무가설")
else:
  print("대립가설")
```
