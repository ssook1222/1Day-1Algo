### 13. Channel 데이터중 2021-10-03일 이후 각 채널의 처음 기록 됐던 구독자 수(subcnt)를 출력하라

### 14. 각 채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

```python
end = channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H') == '2021-11-01 15']
start = channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H') == '2021-10-03 03']
```

`strftime` : 날짜 / 시간을 스트링으로 변환하는 메소드.

```python
end_df = end[['channelname','subcnt']].reset_index(drop=True)
start_df = start[['channelname','subcnt']].reset_index(drop=True)

end_df.columns = ['channelname','end_sub']
start_df.columns = ['channelname','start_sub']
```

```python
df_raw = pd.merge(start_df, end_df)
df_raw['sub_cnt'] = df_raw['end_sub'] - df_raw['start_sub']

df_raw = df_raw[['channelname', 'sub_cnt']]
df_raw
```

### 19. 2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오.

다시 한 번 풀어보기!

```python
start = pd.to_datetime("2021-11-01 00:00:00") # 시작 시간 date time 형태로 변경
end = pd.to_datetime("2021-11-01 15:00:00") # 끝 시간 date time 형태로 변경

target = video.loc[(video["ct"] >= start) & (video['ct'] <= end)].reset_index(drop=True)

def check(x):
    result = max(x) - min(x) # 가장 크게 나온 조회수에서 가장 작게 나온 조회수를 빼서 계산
    return result

answer = target[['videoname','viewcnt']].groupby("videoname").agg(check) # 집계함수로 커스터마이징한 함수 사용
answer
```
