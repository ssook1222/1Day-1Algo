````python
import pandas as pd

DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df = pd.read_csv(DataUrl)
df
````

### 21. quantity컬럼 값이 3인 데이터를 추출하여 첫 5행을 출력하라

````python
df[df['quantity']==3].head(5)
````

### 22. quantity컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력하라

````python
df[df['quantity']==3].reset_index(inplace = False).head(5)
````

`reset_index`를 사용해서 index를 정렬할 수 있음.

### 23. quantity , item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라

````python
df_raw = df[['quantity', 'item_price']]
df_raw
````

### 24. item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라

````python
df_raw['new_price'] = df_raw['item_price'].str[1:].astype('float')
ans = df_raw['new_price'].head()
ans
````

열 전체의 데이터를 변경하고자 할 경우, 열을 가져와서 해당 열에 직접 전처리 한 다음 할당하는 방식도 가능.

### 25. new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라

````python
df_raw[df_raw['new_price'] <=5]['quantity'].count()
````

### 26. item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하라고 index 값을 초기화 하여라

````python
df[df['item_name']=='Chicken Salad Bowl'].reset_index(inplace = False).drop('index', axis=1)
````

#### 열 삭제하기
열 삭제 시 `drop`으로 열 이름을 사용해서 삭제하기


``` python
DataFrame.drop('index', axis=1)
```



### 28. df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화 하여라

````python
df_raw.sort_values(by ='new_price', axis=0).reset_index().drop('index', axis=1)
````

### 29. df의 item_name 컬럼 값중 Chips 포함하는 경우의 데이터를 출력하라

````python
df[df.item_name.str.contains('Chips')]
````

1. `contains('Chips')`를 사용하기 전에는 str로 해당 열의 데이터 타입을 변경해주었는지 먼저 확인을 해주자!

2. 컬럼 값을 DataFrame에서 바로 불러올 수 있음!

### 30. df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하라.

````python
df_raw[df_raw.index%2==0].reset_index().drop('index', axis=1)
````
