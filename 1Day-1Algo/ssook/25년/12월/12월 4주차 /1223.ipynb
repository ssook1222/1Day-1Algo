## BoJ

### 6030

```python
P, Q = map(int, input().split())

# P의 약수 구하기
factors_P = []
for i in range(1, P + 1):
    if P % i == 0:
        factors_P.append(i)

# Q의 약수 구하기
factors_Q = []
for i in range(1, Q + 1):
    if Q % i == 0:
        factors_Q.append(i)

# 모든 조합 출력 (P의 약수, Q의 약수)
for p in factors_P:
    for q in factors_Q:
        print(p, q)
```
